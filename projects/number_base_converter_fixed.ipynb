# Number Base Converter - C++ to Python

Interactive conversions: Binary/Octal/Hex/Decimal/Any Base + Gray Code

```python
def binary_to_decimal(binary):
    return int(binary, 2)

def binary_to_gray(binary):
    gray = binary[0]
    for i in range(1, len(binary)):
        gray += '0' if binary[i-1] == binary[i] else '1'
    return gray

def octal_to_decimal(octal):
    return int(octal, 8)

def decimal_to_binary(dec):
    return bin(dec)[2:] or '0'

def decimal_to_hex(dec):
    return hex(dec)[2:].upper() or '0'

def decimal_to_octal(dec):
    return oct(dec)[2:] or '0'

def decimal_to_any_base(dec, base):
    if not 2 <= base <= 36:
        return "Base out of range!"
    chars = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    result = ''
    d = dec
    while d > 0:
        result = chars[d % base] + result
        d //= base
    return result or '0'

def hex_to_decimal(hex_str):
    return int(hex_str.upper(), 16)

print('Functions defined')
```

```python
# Interactive Menu
while True:
    print("\n----- Number Conversion Menu -----")
    print("1. Binary to Decimal")
    print("2. Binary to Gray")
    print("3. Octal to Decimal")
    print("4. Decimal to Binary")
    print("5. Decimal to Hex")
    print("6. Decimal to Octal")
    print("7. Decimal to Any Base")
    print("8. Hex to Decimal")
    print("0. Exit")
    
    choice = input("Choose: ").strip()
    
    if choice == '0':
        print("Exiting...")
        break
    
    if choice == '1':
        inp = input("Binary: ").strip()
        print("Decimal:", binary_to_decimal(inp))
    elif choice == '2':
        inp = input("Binary: ").strip()
        print("Gray:", binary_to_gray(inp))
    elif choice == '3':
        inp = input("Octal: ").strip()
        print("Decimal:", octal_to_decimal(inp))
    elif choice == '4':
        dec = int(input("Decimal: "))
        print("Binary:", decimal_to_binary(dec))
    elif choice == '5':
        dec = int(input("Decimal: "))
        print("Hex:", decimal_to_hex(dec))
    elif choice == '6':
        dec = int(input("Decimal: "))
        print("Octal:", decimal_to_octal(dec))
    elif choice == '7':
        dec = int(input("Decimal: "))
        base = int(input("Base (2-36): "))
        print("Result:", decimal_to_any_base(dec, base))
    elif choice == '8':
        inp = input("Hex: ").strip()
        print("Decimal:", hex_to_decimal(inp))
    else:
        print("Invalid choice")
```
