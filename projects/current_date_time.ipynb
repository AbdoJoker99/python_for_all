# C++ Current Date/Time → Python

**Exact port** of `<ctime>` + `localtime()` + arrays to Python `datetime` + `strftime`.

Matches:
- `day[]`/`month[]` arrays
- `tm_wday`/`tm_mon`/`tm_mday`/`tm_year+1900`
- `setw(2) setfill('0')` HH:MM:SS

Live + refresh button.

```python
from datetime import datetime
from ipywidgets import interact, Button
import time
```

## Core Functions (C++ tm → Python)

```python
# C++ arrays
month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
         "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
day = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
BASE_YEAR = 1900

def print_date_time():
    now = datetime.now()
    # tm_wday, tm_mon, tm_mday, tm_year + BASE_YEAR
    wday = day[now.weekday()]
    mon = month[now.month - 1]
    mday = now.day
    year = now.year
    
    print(f"Current date: {wday} {mon} {mday} {year}")
    
    # setw(2) setfill('0') HH:MM:SS
    hour = f'{now.hour:02d}'
    minute = f'{now.minute:02d}'
    second = f'{now.second:02d}'
    print(f"Current time: {hour} : {minute} : {second}")
```

## Live Display (Matches C++)

```python
print_date_time()
```

## Interactive Refresh

```python
interact(print_date_time, refresh=Button(description='Refresh'))
```

## Full strftime Alternative (Bonus)

```python
now = datetime.now()
print(now.strftime("%a %b %d %Y"))  # Wed Jan 17 2024
print(now.strftime("%H:%M:%S"))     # 14:30:25
```
