# Hours to Minutes/Seconds Converter - C++ to Python

hrs * 60 = minutes, minutes * 60 = seconds

```python
def hours_to_min_sec(hrs):
    mins = hrs * 60
    secs = mins * 60
    return mins, secs

print('Converter loaded')
```

## Interactive Converter

```python
while True:
    try:
        hrs = float(input("Hours (0 quit): "))
        if hrs == 0: break
        mins, secs = hours_to_min_sec(hrs)
        print(f"{hrs} hours = {mins} minutes = {secs} seconds")
    except:
        print("Float number only")
```

## Tests

```python
test_hrs = [1.5, 2, 0.5]
for h in test_hrs:
    mins, secs = hours_to_min_sec(h)
    print(f"{h} hrs = {mins} min = {secs} sec")
```
