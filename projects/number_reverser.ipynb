# Number Digit Reverser - C++ to Python

1234 → 4321, O(d) digits time. No leading zeros.

```python
def reverse_digits(val):
    num = 0
    temp = val
    while temp != 0:
        num = num * 10 + temp % 10
        temp //= 10
    return num

print('Reverser loaded')
```

## Interactive

```python
while True:
    try:
        n = int(input("Number (0 quit): "))
        if n == 0: break
        rev = reverse_digits(n)
        print(f"{n} reversed is {rev}")
    except:
        print("Integer only")
```

## Tests

```python
tests = [1234, 100, 0, 56]
for t in tests:
    rev = reverse_digits(t)
    print(f"{t} → {rev}")
```
