# Sum of Digits Calculator (C++ → Python)

Ported from C++: Computes sum of digits for any integer (positive/negative).

**Algorithm:** `abs(num)` → while num > 0: sum += num % 10; num //= 10  
**Time:** O(digits)

Interactive widget for testing.

```python
from ipywidgets import interact, IntText
import math

def sum_of_digits(num):
    """Sum of digits, handles negative via abs."""
    total = 0
    num = abs(num)
    while num > 0:
        total += num % 10
        num //= 10
    return total
```

```python
# Interactive tester
@interact(num=IntText(value=12345, description='Number:'))
def test_sum(num):
    result = sum_of_digits(num)
    print(f"Sum of digits of {num} = {result}")
    print(f"C++ equivalent output: Sum of digits = {result}")
```

```python
# Test cases
test_cases = [123, -456, 0, 999, 7]
for n in test_cases:
    print(f'{n}: {sum_of_digits(n)}')
```

## Manual Input (like C++ main)
