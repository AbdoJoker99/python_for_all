## String Operations

```python
name = "world"
formatted = f"Hello, {name}!"
print("Formatted:", formatted)

s = "hello"
sub = s[0:5]
print("Substring:", sub)

s = s.ljust(8)
print("Padded : '", s, "'")

words = ["hello", "world", "java"]
joined_str = ', '.join(words)
print("Joined:", joined_str)

last_index = s.rfind("l")
print("Last 'l' index:", last_index)

print("Length:", len(s))

replaced = s.replace('h', 'H')
print("Replaced h->H:", replaced)
```

```python
# Split function
def split(str, delim):
    return str.split(delim)

splitted = split("a,b,c", ',')
print("Splitted:", splitted)

print("Upper:", s.upper())
print("Lower:", s.lower())
print("Reversed:", s[::-1])
```

## Math Functions (import math)

```python
import math

print(f"acos(0.5): {math.acos(0.5):.2f}")
print(f"asin(0.5): {math.asin(0.5):.2f}")
print(f"atan(1): {math.atan(1):.2f}")
print(f"Cube root 27: {27**(1/3):.2f}")
print(f"ceil(3.7): {math.ceil(3.7)}")
print(f"cosh(1): {math.cosh(1):.2f}")
print(f"exp(1): {math.exp(1):.2f}")
print(f"floor(3.7): {math.floor(3.7)}")
print(f"fmod(17,5): {math.fmod(17,5)}")
print(f"log(e): {math.log(math.e):.2f}")
print(f"log10(100): {math.log10(100)}")
print(f"max(5,10): {max(5,10)}")
print(f"min(5,10): {min(5,10)}")
```

## Array/List Operations

```python
arr = [1, 2, 3, 4, 5]
new_arr = arr[1:3]
print("Sliced (1:3):", new_arr)

print("Equal:", arr == [1,2,3,4,5])

filled = [5] * 5
print("Filled 5s:", filled)

arr_sorted = sorted(arr)
print("Sorted asc:", arr_sorted)

arr_desc = sorted(arr, reverse=True)
print("Sorted desc:", arr_desc)
```

## String Utilities

```python
txt = "     banana     ".strip()
print("Trimmed : '", txt, "'")

txt2 = "....banana....".strip('.,grt')
print("Trim(.,grt): '", txt2, "'")

print("Starts 'banana':", txt.startswith('banana'))
print("Ends 'a':", txt.endswith('a'))

print("Find 'banana':", txt.find('banana'))
print("Find 'q':", txt.find('q'))

print("All lower:", txt.islower())
print("All alpha:", txt.isalpha())
```

## Set Operations

```python
fruits = {'apple', 'banana'}
fruits.add('orange')
print("Fruits set:", fruits)

x = {'apple'}
y = {'google'}

disjoint = x.isdisjoint(y)
print("Disjoint:", disjoint)

union_set = x | y
print("Union:", union_set)
```

## Binary/Octal/Hex

```python
print("Binary 36:", bin(36)[2:].zfill(8))
print("Octal 12:", oct(12))
print("Hex 255:", hex(255))
```

## Char to Int (ord)

```python
val = ord('h')
print("ASCII 'h':", val)
```
