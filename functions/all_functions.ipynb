```python
def greet_user():
    """Display a simple greeting."""
    print("Hello!")

greet_user()
```

```python
def greet_user(username):
    print(f"Hello, {username}!")

greet_user('jesse')
```

```python
def describe_pet(animal, name='dog'):
    print(f"I have a {animal}.")
    print(f"My {animal}'s name is {name.title()}." )

describe_pet('hamster', name='harry')
describe_pet(name='willie')
```

```python
def make_pizza(topping='bacon'):
    print(f"Have a {topping} pizza!")

make_pizza()
make_pizza('pepperoni')
```

```python
def add_numbers(x, y):
    return x + y

print(add_numbers(3, 5))
```

```python
def build_person(first, last, age=None):
    person = {'first': first, 'last': last}
    if age:
        person['age'] = age
    return person

print(build_person('jimi', 'hendrix', 27))
```

## Lists as Arguments

```python
def print_models(unprinted, printed):
    while unprinted:
        current = unprinted.pop()
        print(f"Printing {current}")
        printed.append(current)

unprinted = ['phone case', 'pendant', 'ring']
printed = []
print_models(unprinted[:], printed)  # Copy to protect original
print('Original:', unprinted)
```

## *args, **kwargs

```python
def make_pizza(size, *toppings):
    print(f"Making {size}-inch pizza with toppings: {toppings}")

make_pizza('small', 'pepperoni', 'bacon')
```

```python
def build_profile(first, last, **info):
    profile = {'first': first, 'last': last}
    profile.update(info)
    return profile

print(build_profile('albert', 'einstein', location='princeton'))
```

## Class Methods (oop/python_classes.ipynb)

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.fuel_capacity = 15
        self.fuel_level = 0

    def fill_tank(self):
        self.fuel_level = self.fuel_capacity
        print("Fuel tank full.")

    def drive(self):
        print("Driving.")

    def update_fuel_level(self, new_level):
        if new_level <= self.fuel_capacity:
            self.fuel_level = new_level
        else:
            print("Too much!")

my_car = Car('audi', 'a4', 2016)
my_car.fill_tank()
my_car.drive()
```

## Functions from exceptions/python_exceptions.ipynb

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age negative")
    if age < 18:
        raise PermissionError("Too young")
    print("Access OK")

try:
    check_age(15)
except (ValueError, PermissionError) as e:
    print(e)
```
