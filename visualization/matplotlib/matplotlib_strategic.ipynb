## 2. 'Less is More' Declutter (Image 2)

```python
months = np.arange(1,13)
product_a = 100 + np.cumsum(np.random.randn(12)*3)
product_b = 110 + np.cumsum(np.random.randn(12)*2)
product_c = 90 + np.cumsum(np.random.randn(12)*5)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))

ax1.plot(months, product_a, 'b-', label='A', alpha=0.5, lw=1)
ax1.plot(months, product_b, 'r--', label='B', alpha=0.5, lw=1)
ax1.plot(months, product_c, 'g-.', label='C', lw=1)
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.8)
ax1.set_title('WITH CLUTTER')

ax2.plot(months, product_a, color='lightgray', alpha=0.5, lw=1)
ax2.plot(months, product_b, color='gray', alpha=0.3, lw=1)
ax2.plot(months, product_c, color='darkgreen', lw=3, label='Product C')
ax2.text(10, product_c[-1], 'C: Growth!', fontsize=12, color='darkgreen')
ax2.set_title('WITHOUT CLUTTER - Focus: C Growth')

plt.tight_layout()
plt.show()
```

## 3. Comparative Box Plots (Image 3)

```python
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
data = [np.random.normal(100 + i*10, 15, 200) for i in range(4)]

fig, ax = plt.subplots(figsize=(10,6))
box = ax.boxplot(data, labels=categories, patch_artist=True, boxprops=dict(facecolor='lightblue'))

ax.set_title('Category Comparison')
ax.set_ylabel('Values')
ax.grid(True, alpha=0.3)
plt.show()
```

## 4. Correlation Analysis (Image 4)

```python
np.random.seed(42)
ad_budget = np.random.uniform(10, 100, 50)
sales = 20 + 0.8 * ad_budget + np.random.normal(0, 10, 50)

plt.figure(figsize=(10,7))
plt.scatter(ad_budget, sales, s=60, alpha=0.7, c='coral', edgecolors='black')
plt.plot([10,100], [20+0.8*10, 20+0.8*100], 'r--', lw=2, label='Trend')
plt.xlabel('Ad Budget ($k)')
plt.ylabel('Sales ($k)')
plt.title('Ad Budget vs Sales Correlation')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

corr = np.corrcoef(ad_budget, sales)[0,1]
print(f'Correlation coefficient: {corr:.3f}')
```

## Real-World: Healthcare

```python
departments = ['ER', 'Cardio', 'Pediatrics']
wait_times = [np.random.exponential(20, 200) for _ in departments]

plt.figure(figsize=(8,6))
box = plt.boxplot(wait_times, labels=departments)
plt.title('Patient Wait Times (minutes)')
plt.ylabel('Minutes')
plt.show()
```

## Interactive Declutter

```python
from ipywidgets import interact

@interact(highlight=(0,3), alpha=(0.1,0.9,0.1))
def declutter(highlight=2, alpha=0.3):
    plt.figure(figsize=(12,6))
    months = np.arange(12)
    for i in range(4):
        y = 100 + i*10 + np.cumsum(np.random.randn(12)*2)
        color = 'darkblue' if i == highlight else 'lightgray'
        plt.plot(months, y, color=color, alpha=alpha if i != highlight else 1, lw=2)
    plt.title('Declutter Demo')
    plt.show()
```
