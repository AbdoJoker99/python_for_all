```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('default')
```

## 1. Setup & Foundations (Images 1-2)

```python
# Basic scatter (0-1 range, scale *100)
x = np.random.random(50) * 100
y = np.random.random(50) * 100
plt.scatter(x, y)
plt.show()
```

## 2. Visualizing (Image 3)

```python
# plt.show() explicit (for IDEs)
plt.figure(figsize=(6,4))
plt.scatter(x[:20], y[:20], color='blue')
plt.show()
```

## 3. Advanced Customization (Images 4-8)

```python
# Markers, colors, size, alpha (overlap density)
plt.figure(figsize=(10,8))

plt.scatter(x, y, marker='*', color='#0f0', s=100, alpha=0.5)  # stars, green hex, size, transparency
plt.title('Custom Scatter: Markers + Alpha')
plt.show()
```

## 4. Time Series Line Plots (Images 10-11)

```python
# Weight tracking (2006-2021)
years = np.arange(2006, 2022)
weights = 70 + np.cumsum(np.random.randn(16) * 0.5)  # trend + noise

plt.figure(figsize=(10,6))
plt.plot(years, weights, 'b-o', linewidth=2, markersize=6)
plt.title('Weight Tracking Over Time')
plt.grid(True, alpha=0.3)
plt.show()
```

## 5. Labels, Legends, Subplots (Next Steps)

```python
# Multiple datasets + labels/legend
plt.figure(figsize=(12,8))

plt.scatter(x, y, alpha=0.6, label='Random Data', s=60)
plt.plot(years, weights, 'r-', label='Weight Trend', linewidth=2)

plt.title('Combined Plot: Scatter + Line')
plt.xlabel('X / Years')
plt.ylabel('Y / Weight')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

ax1.scatter(x, y, alpha=0.5)
ax1.set_title('Scatter')

ax2.plot(years, weights, 'g--')
ax2.set_title('Line')

plt.tight_layout()
plt.show()
```

## 6. Advanced Techniques

```python
# Histograms (density)
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(x, bins=20, alpha=0.7, color='skyblue')
plt.title('X Distribution')

plt.subplot(1,2,2)
plt.hist(weights, bins=10, alpha=0.7, color='lightgreen')
plt.title('Weight Distribution')

plt.tight_layout()
plt.show()
```

## Interactive Widgets

```python
from ipywidgets import interact

@interact(marker=['o','*','s','^'], color=['blue','green','#ff69b4'], alpha=(0.1,1.0,0.1), size=(20,200))
def interactive_scatter(marker='o', color='blue', alpha=0.6, size=60):
    plt.figure(figsize=(8,6))
    plt.scatter(x, y, marker=marker, c=color, s=size, alpha=alpha)
    plt.title(f'Interactive: {marker} {color} α={alpha} s={size}')
    plt.show()
```
