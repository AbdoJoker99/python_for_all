```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

## 1. Polishing & Exporting (Images 1-3)

```python
# Subplots with suptitle + savefig
fig, axs = plt.subplots(1, 2, figsize=(12,5))

axs[0].plot(np.sin(np.linspace(0,10,100)))
axs[0].set_title('Sine')

axs[1].plot(np.cos(np.linspace(0,10,100)))
axs[1].set_title('Cosine')

fig.suptitle('Trigonometric Functions', fontsize=16, y=0.95)

# Export pro
plt.tight_layout()
plt.savefig('trig_functions.png', dpi=300, bbox_inches='tight', transparent=False)
plt.savefig('trig_functions.pdf', bbox_inches='tight')
plt.show()
print('Saved: trig_functions.png/pdf (300dpi, tight bbox)')
```

## 2. 3D Foundations (Images 4-7)

```python
# 3D scatter + line
fig = plt.figure(figsize=(12,5))

# 3D scatter
ax1 = fig.add_subplot(121, projection='3d')
x, y, z = np.random.random((3,50)) * 10
ax1.scatter(x, y, z, c=z, cmap='viridis')
ax1.set_title('3D Scatter')

# 3D line
ax2 = fig.add_subplot(122, projection='3d')
t = np.linspace(0, 10, 100)
ax2.plot(np.sin(t), np.cos(t), t)
ax2.set_title('3D Helix')

plt.show()
```

## 3. Surface Plots (Images 8-9)

```python
# Meshgrid + surface
X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='Spectral', alpha=0.8)
ax.set_title('3D Surface: sin(√(x²+y²))')
fig.colorbar(surf, shrink=0.5)

plt.show()
```

## Interactive 3D Surface

```python
from ipywidgets import interact

def surface_demo(func='sin', cmap='Spectral'):
    functions = {
        'sin': lambda r: np.sin(r),
        'cos': lambda r: np.cos(r),
        'exp': lambda r: np.exp(-r/5)
    }
    f = functions[func]
    
    X, Y = np.meshgrid(np.linspace(-5, 5, 30), np.linspace(-5, 5, 30))
    R = np.sqrt(X**2 + Y**2)
    Z = f(R)
    
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap=cmap, alpha=0.9)
    ax.set_title(f'3D Surface: {func}(r)')
    plt.show()

interact(surface_demo, func=['sin','cos','exp'], cmap=['viridis','Spectral','plasma'])
```

## Export Showcase

```python
# Different export modes
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(np.sin(np.linspace(0,10,100)), 'b-', linewidth=3)
ax.set_title('Export Test')

# PNG high-res
plt.savefig('high_res.png', dpi=300, bbox_inches='tight', facecolor='white')
# Transparent
plt.savefig('transparent.png', transparent=True, bbox_inches='tight')
# PDF vector
plt.savefig('vector.pdf', bbox_inches='tight')
plt.show()

print('Exported: high_res.png | transparent.png | vector.pdf')
```

## Visualization Principles (Image 11)

**4 Core Principles:**

1. **Simplicity** - Remove clutter (grid alpha=0.3, tight bbox)
2. **Clarity** - Labels legible, legends positioned (`loc='best'`)
3. **Accuracy** - Proper scales, no distortion (aspect='equal')
4. **Efficiency** - Convey message fast (color=cmap by value)

**Goal:** Audience grasps insight in <5 seconds.

**Avoid:** 3D unless necessary (distorts), too many colors.

## Complete Export Workflow

```python
# Pro publication-ready plot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14,12))

t = np.linspace(0, 4*np.pi, 100)
ax1.plot(t, np.sin(t), 'b-', lw=2)
ax1.set_title('Sine')

X, Y = np.meshgrid(np.linspace(-3,3,30), np.linspace(-3,3,30))
ax2 = fig.add_subplot(2,2,2, projection='3d')
ax2.plot_surface(X, Y, np.sin(np.sqrt(X**2+Y**2)), cmap='coolwarm')

# Publication save
plt.tight_layout()
fig.suptitle('Matplotlib Master Dashboard', fontsize=16, y=0.98)
plt.savefig('master_dashboard.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.savefig('master_dashboard.svg', format='svg', bbox_inches='tight')  # Vector
plt.show()
print('🏆 Publication-ready: PNG 300dpi + SVG vector')
```
