```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Line Style Customization (Image 1)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(12,4))
plt.plot(x, y1, 'b-', label='sin (solid blue)', linewidth=2)
plt.plot(x, y2, 'r--', label='cos (dashed red)', linewidth=2)
plt.legend()
plt.title('Line Styles: "color-line"')
plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Distribution Analysis (Images 3-5)

```python
# Histograms + bins + cumulative
data = np.random.normal(50, 15, 1000)  # age-like data

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,4))

# Standard hist
ax1.hist(data, bins=20, color='skyblue', edgecolor='black')
ax1.set_title('Standard Histogram (20 bins)')

# Custom bins (age groups)
bins = [0,18,21,30,40,50,65,100]
ax2.hist(data, bins=bins, color='lightgreen')
ax2.set_title('Age Groups')

# Cumulative
ax3.hist(data, bins=30, cumulative=True, density=True, color='orange', alpha=0.7)
ax3.set_title('Cumulative Distribution')

plt.tight_layout()
plt.show()
```

## 3. Categorical Data (Images 6-8)

```python
# Bar chart
langs = ['Python', 'JavaScript', 'Java', 'C#', 'C++']
popularity = [85, 78, 65, 55, 48]

plt.figure(figsize=(10,6))
plt.bar(langs, popularity, color=['#ff6b6b', '#4ecdc4', '#45b7d1', '#f9ca24', '#f0932b'])
plt.title('Programming Language Popularity')
plt.ylabel('Popularity %')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart + explode (Image 8)
langs_short = ['Python', 'JS', 'Java', 'C#']
sizes = [35, 25, 20, 20]
explode = (0.1, 0, 0, 0)  # explode Python

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=langs_short, explode=explode, autopct='%1.1f%%',
        pctdistance=0.85, shadow=True, startangle=90)
plt.title('Language Share (C# exploded)')
plt.show()
```

## 4. Boxplots (Images 9-10)

```python
# Box & whisker (outliers)
np.random.seed(42)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(110, 15, 200)
data3 = np.concatenate([np.random.normal(90, 5, 180), [50, 160, 170]])  # outliers

data = [data1, data2, data3]

plt.figure(figsize=(10,6))
box = plt.boxplot(data, labels=['Group A', 'Group B', 'Group C with outliers'])
plt.title('Box & Whisker Plots')
plt.ylabel('Values')
plt.grid(True, axis='y', alpha=0.3)
plt.show()
```

## Interactive Demos

```python
from ipywidgets import interact

@interact(line_style=['-', '--', '-.', ':'], color=['r','g','b','orange'], bins=(5,50,5))
def interactive_hist(line_style='-', color='blue', bins=20):
    data = np.random.normal(50, 15, 1000)
    plt.figure(figsize=(10,6))
    plt.hist(data, bins=bins, histtype='step', linewidth=2,
             linestyle=line_style, color=color)
    plt.title(f'Interactive Histogram: {bins} bins')
    plt.show()
```

## Combined Statistical Dashboard

```python
fig = plt.figure(figsize=(15,10))

# Hist + box + bar
data_normal = np.random.normal(50, 15, 1000)

plt.subplot(2,3,1)
plt.hist(data_normal, bins=30)
plt.title('Histogram')

plt.subplot(2,3,2)
plt.boxplot([data_normal])
plt.title('Boxplot')

plt.subplot(2,3,3)
langs = ['Python', 'JS']
plt.pie([60, 40], labels=langs, autopct='%1.1f%%')
plt.title('Pie')

plt.tight_layout()
plt.show()
```
