```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Professional Annotation (Images 2-3)

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10,6))
plt.plot(x, y, 'b-', linewidth=2)

# Labels + custom font
plt.title('Sine Wave', fontsize=16, fontname='serif', pad=20)
plt.xlabel('Time (s)', fontsize=12, fontname='Arial')
plt.ylabel('Amplitude', fontsize=12)

# Custom ticks (k USD style)
ticks = [0, 2, 4, 6, 8, 10]
labels = ['0k', '2k', '4k', '6k', '8k', '10k']
plt.xticks(ticks, labels)

plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Multi-Dataset Comparison (Images 5-7)

```python
# Stock overlays A/B/C
days = np.arange(1, 31)
stock_a = 100 + np.cumsum(np.random.randn(30) * 2)
stock_b = 100 + np.cumsum(np.random.randn(30) * 1.5)
stock_c = 100 + np.cumsum(np.random.randn(30) * 2.5)

plt.figure(figsize=(12,7))
plt.plot(days, stock_a, 'b-', label='Stock A', linewidth=2, alpha=0.8)
plt.plot(days, stock_b, 'r--', label='Stock B', linewidth=2, alpha=0.8)
plt.plot(days, stock_c, 'g-.', label='Stock C', linewidth=2, alpha=0.8)

plt.title('Stock Performance Comparison', fontsize=14, pad=15)
plt.xlabel('Days')
plt.ylabel('Price ($)')
plt.legend(loc='lower center', framealpha=0.9)
plt.grid(True, alpha=0.3)
plt.show()
```

## 3. Advanced Subplots & Grids (Images 8-10)

```python
# 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(12,10))
x = np.linspace(0, 4*np.pi, 1000)

# Top left [0,0]
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sin(x)')

# Top right [0,1]
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cos(x)')

# Bottom left [1,0]
axs[1, 0].plot(x, np.log(np.abs(x)+1))
axs[1, 0].set_title('log(|x|+1)')

# Bottom right [1,1]
axs[1, 1].plot(x, np.sin(x)*np.cos(x))
axs[1, 1].set_title('sin(x)cos(x)')

plt.tight_layout()
plt.show()
```

## Interactive Multi-Plot

```python
from ipywidgets import interact

@interact(n_lines=(1,5), style=['-', '--', '-.', ':'])
def multi_stock(n_lines, style='-'):
    days = np.arange(1, 31)
    plt.figure(figsize=(12,7))
    colors = plt.cm.Set1(np.linspace(0,1,n_lines))
    
    for i in range(n_lines):
        data = 100 + np.cumsum(np.random.randn(30)*2)
        plt.plot(days, data, color=colors[i], ls=style, label=f'Stock {chr(65+i)}',
                linewidth=2, alpha=0.8)
    
    plt.title(f'Multi-Stock Overlay ({n_lines} lines)')
    plt.xlabel('Days')
    plt.ylabel('Price ($)')
    plt.legend(loc='best')
    plt.grid(True, alpha=0.3)
    plt.show()
```

## Pro Dashboard (All Together)

```python
fig = plt.figure(figsize=(16,12))

# 4 pro plots
gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3)

# Annotated sin
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(np.linspace(0,10,100), np.sin(np.linspace(0,10,100)))
ax1.set_title('Annotated Sine', fontname='serif', fontsize=14)
ax1.set_xticks([0,5,10], ['Start', 'Mid', 'End'])

# Multi-line stocks
ax2 = fig.add_subplot(gs[0,1])
days = np.arange(30)
for i, label in enumerate(['A','B','C']):
    data = 100 + np.cumsum(np.random.randn(30)*1.5)
    ax2.plot(days, data, label=label)
ax2.legend(loc='lower center')
ax2.set_title('Stocks Dashboard')

plt.show()
```
