# Pygal Advanced - Styles, Config, Maps

Styling, config, world maps (`pip install pygal_maps_world`).

## Built-in Styles

```python
import pygal
from pygal.style import LightGreenStyle
x_values = list(range(11))
squares = [x**2 for x in x_values]
cubes = [x**3 for x in x_values]
chart_style = LightGreenStyle()
chart = pygal.Line(style=chart_style)
chart.force_uri_protocol = 'http'
chart.title = "Squares and Cubes"
chart.x_labels = x_values
chart.add('Squares', squares)
chart.add('Cubes', cubes)
chart.render_to_file('styled.svg')
```

## Parametric Styles

```python
from pygal.style import LightenStyle
chart_style = LightenStyle('#336688')
chart = pygal.Line(style=chart_style)
chart.force_uri_protocol = 'http'
chart.title = "Custom Color"
chart.add('Data', squares)
chart.render_to_file('lighten.svg')
```

## Custom Style Properties

```python
chart_style = LightenStyle('#336688')
chart_style.plot_background = '#CCCCCC'
chart_style.major_label_font_size = 20
chart_style.label_font_size = 16
chart = pygal.Line(style=chart_style)
chart.add('Data', squares)
chart.render_to_file('custom_style.svg')
```

## Bare Style Class

```python
from pygal.style import Style
chart_style = Style()
chart_style.colors = ['#CCCCCC', '#AAAAAA', '#888888']
chart_style.plot_background = '#EEEEEE'
chart = pygal.Line(style=chart_style)
chart.force_uri_protocol = 'http'
chart.add('Data1', squares)
chart.add('Data2', cubes)
chart.render_to_file('bare_style.svg')
```

## Config Object

```python
my_config = pygal.Config()
my_config.show_y_guides = False
my_config.width = 1000
my_config.dots_size = 5
chart = pygal.Line(config=my_config)
chart.force_uri_protocol = 'http'
chart.add('Data', squares)
chart.render_to_file('config.svg')
```

## Styling Series / Points

```python
chart = pygal.Line()
chart.force_uri_protocol = 'http'
chart.add('Squares', squares, dots_size=2)
chart.add('Cubes', cubes, dots_size=3)
chart.render_to_file('series_style.svg')

# Individual points
repos = [
    {'value': 20506, 'color': '#3333CC', 'xlink': 'http://djangoproject.com/'},
    20054, 12607, 11827
]
chart = pygal.Bar()
chart.force_uri_protocol = 'http'
chart.x_labels = ['django', 'requests', 'scikit-learn', 'tornado']
chart.y_title = 'Stars'
chart.add('Python Repos', repos)
chart.render_to_file('repos.svg')
```

## World Maps (`pip install pygal_maps_world`)

```python
from pygal.maps.world import World
wm = World()
wm.force_uri_protocol = 'http'
wm.title = 'North America'
wm.add('NA', ['ca', 'mx', 'us'])
wm.render_to_file('na.svg')

# Population data
populations = {'ca': 34126000, 'us': 309349000, 'mx': 113423000}
wm = World()
wm.force_uri_protocol = 'http'
wm.title = 'NA Populations'
wm.add('Pop', populations)
wm.render_to_file('na_pop.svg')

# Country codes
# from pygal.maps.world import COUNTRIES
# print(sorted(COUNTRIES.items()))
```

## Summary
- Styles: `LightGreenStyle()`, `LightenStyle(color)`
- Custom: `Style()`, set colors/background/font
- Config: `pygal.Config()` width/dots etc.
- Series/points: `dots_size`, dicts with value/color/xlink
- Maps: `World()`, country codes ['us'], dict data
