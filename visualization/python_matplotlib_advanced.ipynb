# Matplotlib Advanced Cheat Sheet

Additional features: multiple data sets, fill_between, datetime, subplots, more.

## Plotting Two Data Sets

```python
import matplotlib.pyplot as plt
x_values = list(range(11))
squares = [x**2 for x in x_values]
cubes = [x**3 for x in x_values]
plt.scatter(x_values, squares, c='blue', edgecolor='none', s=20)
plt.scatter(x_values, cubes, c='red', edgecolor='none', s=20)
plt.axis([0, 11, 0, 1100])
plt.show()
```

## Fill Between Data Sets

```python
plt.fill_between(x_values, cubes, squares, facecolor='blue', alpha=0.25)
plt.show()
```

## Datetime Formatting
%A Weekday, %B Month, %m Month num, %d Day, %Y Year4, %y Year2, %H Hour24, %I Hour12, %p AM/PM, %M Min, %S Sec

```python
from datetime import datetime as dt
new_years = dt.strptime('1/1/2017', '%m/%d/%Y')
ny_string = dt.strftime(new_years, '%B %d, %Y')
print(ny_string)
```

## Plotting Dates (High Temps)

```python
from matplotlib import dates as mdates
dates = [dt(2016, 6, 21), dt(2016, 6, 22), dt(2016, 6, 23), dt(2016, 6, 24)]
highs = [57, 68, 64, 59]
plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red')
plt.title("Daily High Temps", fontsize=24)
plt.ylabel("Temp (F)", fontsize=16)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B %d %Y'))
plt.gcf().autofmt_xdate()
plt.show()
```

## Subplots - Shared X

```python
x_vals = list(range(11))
squares = [x**2 for x in x_vals]
cubes = [x**3 for x in x_vals]
fig, axarr = plt.subplots(2, 1, sharex=True)
axarr[0].scatter(x_vals, squares)
axarr[0].set_title('Squares')
axarr[1].scatter(x_vals, cubes, c='red')
axarr[1].set_title('Cubes')
plt.show()
```

## Subplots - Shared Y

```python
fig, axarr = plt.subplots(1, 2, sharey=True)
axarr[0].scatter(x_vals, squares)
axarr[0].set_title('Squares')
axarr[1].scatter(x_vals, cubes, c='red')
axarr[1].set_title('Cubes')
plt.show()
```

## Datetime Objects

```python
today = dt.now()
print(dt.strftime(today, '%m/%d/%Y'))
new_years = dt(2017, 1, 1)
fall_equinox = dt(year=2016, month=9, day=22)
print(new_years, fall_equinox)
```

## Summary
- Multiple plots: call scatter/plot multiple times
- `fill_between(x, y1, y2)`
- `datetime.strptime(str, fmt)`, `strftime(dt, fmt)`
- `plt.subplots(rows, cols, sharex/y=True)`
- Date plots: `mdates.DateFormatter`, `autofmt_xdate()`
