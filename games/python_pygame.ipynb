## Basic Window & Loop

```python
import sys
import pygame as pg
pg.init()
screen = pg.display.set_mode((1200, 800))
pg.display.set_caption("Game")
bg_color = (230, 230, 230)
screen.fill(bg_color)
pg.display.flip()
```

## Rect Objects & Attributes

```python
screen_rect = screen.get_rect()
print('left:', screen_rect.left, 'right:', screen_rect.right)
print('top:', screen_rect.top, 'bottom:', screen_rect.bottom)
print('center:', screen_rect.centerx, screen_rect.centery)
print('size:', screen_rect.size)
# Create bullet rect
bullet_rect = pg.Rect(100, 100, 3, 15)
pg.draw.rect(screen, (100,100,100), bullet_rect)
```

## Text Rendering

```python
msg = "Play again?"
msg_color = (100, 100, 100)
f = pg.font.SysFont(None, 48)
msg_image = f.render(msg, True, msg_color)
msg_rect = msg_image.get_rect()
msg_rect.center = screen_rect.center
screen.blit(msg_image, msg_rect)
pg.display.flip()
```
