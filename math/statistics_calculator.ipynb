# 📊 Statistical Calculator (C++ → Python Port)

**Exact equivalent** of provided C++ code.

**Features:** Mean, Median, Mode (multi-mode support), Variance, StdDev

**Time:** O(n) mean/mode/var, O(n log n) median

Interactive + your sample data + custom input.

```python
from collections import Counter
import math
from typing import List
```

## Core Functions (Exact C++ Match)

```python
def calculate_mean(data: List[int]) -> float:
    """O(n) - C++ calculateMean"""
    return sum(data) / len(data)

def calculate_median(data: List[int]) -> float:
    """O(n log n) - C++ calculateMedian"""
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        return (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    return sorted_data[n//2]

def calculate_mode(data: List[int]) -> List[int]:
    """O(n) - C++ calculateMode (multi-mode)"""
    freq = Counter(data)
    if not freq:
        return []
    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]
    # No mode if all unique
    if len(modes) == len(data):
        return []
    return modes

def calculate_variance(data: List[int], mean: float) -> float:
    """O(n) - Population variance"""
    return sum((x - mean) ** 2 for x in data) / len(data)

def calculate_std_dev(variance: float) -> float:
    """O(1)"""
    return math.sqrt(variance)
```

## Your Sample Data Test

```python
data = [1, 2, 2, 3, 4, 4, 4, 5, 6]

print('Data:', data)

mean = calculate_mean(data)
median = calculate_median(data)
mode = calculate_mode(data)
variance = calculate_variance(data, mean)
std_dev = calculate_std_dev(variance)

print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Mode:', mode if mode else 'No mode')
print(f'Variance: {variance}')
print(f'Standard Deviation: {std_dev}')
```

## Interactive Calculator

```python
from ipywidgets import interact, IntText

def stats_demo(*numbers):
    data = [n for n in numbers if n is not None]
    if not data:
        print('Enter numbers')
        return
    
    mean = calculate_mean(data)
    median = calculate_median(data)
    mode = calculate_mode(data)
    variance = calculate_variance(data, mean)
    std_dev = calculate_std_dev(variance)
    
    print(f'Data: {data}')
    print(f'Mean: {mean:.2f}')
    print(f'Median: {median:.2f}')
    print(f'Mode: {mode if mode else "No mode"}')
    print(f'Variance: {variance:.4f}')
    print(f'Std Dev: {std_dev:.4f}')

interact(stats_demo,
         n1=IntText(1,value=1),
         n2=IntText(2,value=2),
         n3=IntText(2,value=2),
         n4=IntText(3,value=3),
         n5=IntText(4,value=4),
         n6=IntText(4,value=4),
         n7=IntText(4,value=4),
         n8=IntText(5,value=5),
         n9=IntText(6,value=6))
```

## Bulk Data Input

```python
def analyze_dataset(data_str: str):
    data = [int(x) for x in data_str.split(',')]
    mean = calculate_mean(data)
    median = calculate_median(data)
    mode = calculate_mode(data)
    variance = calculate_variance(data, mean)
    std_dev = calculate_std_dev(variance)
    
    print(f'Dataset: {data} (n={len(data)})\n')
    print(f'Mean:           {mean:8.3f}')
    print(f'Median:         {median:8.3f}')
    print(f'Mode:           {mode or "None"}')
    print(f'Variance:       {variance:8.4f}')
    print(f'Std Deviation:  {std_dev:8.4f}')

# Test different datasets
analyze_dataset('1,3,3,6,7,8,9')
print('-' * 40)
analyze_dataset('1,2,3,4,5')  # uniform
print('-' * 40)
analyze_dataset('10,10,10,10') # all same
```
