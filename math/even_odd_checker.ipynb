# Even/Odd Checker - C++ to Python

Simple %2 modulus check with interactive input.

```python
def is_even_odd(number):
    remainder = number % 2
    if remainder == 0:
        return f"{number} is even"
    else:
        return f"{number} is odd"

print('Checker loaded')
```

## Interactive

```python
while True:
    try:
        n = int(input("Number (0 quit): "))
        if n == 0: break
        print(is_even_odd(n))
    except:
        print("Integer only")
```

## Tests

```python
tests = [-4, 0, 1, 7, 8]
for t in tests:
    print(is_even_odd(t))
```
