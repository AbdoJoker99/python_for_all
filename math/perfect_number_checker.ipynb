# Perfect Number Checker - C++ to Python

**O(n) time** - sum proper divisors == number (e.g., 6: 1+2+3=6)

```python
def is_perfect(num):
    if num <= 1:
        return False
    sum_div = 0
    for i in range(1, num):
        if num % i == 0:
            sum_div += i
    return sum_div == num

print('Checker loaded')
```

## Interactive

```python
while True:
    try:
        n = int(input("Number (0 quit): "))
        if n == 0: break
        status = "perfect" if is_perfect(n) else "not perfect"
        print(f"{n} is {status}")
    except:
        print("Integer only")
```

## Tests (known perfect: 6, 28, 496)

```python
tests = [6, 28, 496, 1, 10, 25]
for t in tests:
    status = "perfect" if is_perfect(t) else "not perfect"
    print(f"{t}: {status}")
```
