# Recursive Even/Odd Sort - C++ to Python

Separate even/odd, recursive insertion sort, merge evens+odds. Input [3,1,2,4,7,6] → [2,4,6,1,3,7]

```python
def insert_sorted(arr, val):
    if not arr or arr[-1] <= val:
        arr.append(val)
        return
    temp = arr.pop()
    insert_sorted(arr, val)
    arr.append(temp)

def sort_recursive(arr):
    if len(arr) <= 1:
        return
    temp = arr.pop()
    sort_recursive(arr)
    insert_sorted(arr, temp)

def sort_even_odd_recursive(input_list):
    evens, odds = [], []
    def recurse(index):
        if index >= len(input_list):
            return
        if input_list[index] % 2 == 0:
            evens.append(input_list[index])
        else:
            odds.append(input_list[index])
        recurse(index + 1)
    recurse(0)
    sort_recursive(evens)
    sort_recursive(odds)
    return evens + odds

print('Sorter loaded')
```

## Test (matches C++)

```python
input_arr = [3, 1, 2, 4, 7, 6]
result = sort_even_odd_recursive(input_arr)
print("Input:", input_arr)
print("Sorted (even then odd):", result)
```
