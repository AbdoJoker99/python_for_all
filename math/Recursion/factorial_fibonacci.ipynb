# Factorial & Fibonacci - C++ to Python

Recursive factorial O(n), recursive fib O(2^n), iterative fib O(n)

```python
def factorial_recursive(n):
    return 1 if n <= 1 else n * factorial_recursive(n-1)

def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

def fibonacci_iterative(n):
    if n <= 0: return []
    fib = [0]
    if n == 1: return fib
    fib.append(1)
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib

print('Functions loaded')
```

## Demos (n=5 matches C++)

```python
num = 5
print(f"Factorial({num}): {factorial_recursive(num)}")
print(f"Fib recursive({num}): {fibonacci_recursive(num)}")
print(f"Fib iterative({num}): {fibonacci_iterative(num)}")
```

## Interactive

```python
while True:
    try:
        n = int(input("n (0 quit): "))
        if n == 0: break
        print(f"fact({n}): {factorial_recursive(n)}")
        print(f"fib({n}): {fibonacci_recursive(n)}, seq: {fibonacci_iterative(n)}")
    except:
        print("Non-negative integer")
```
