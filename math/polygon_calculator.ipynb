# Polygon Area & Perimeter Calculator - C++ to Python

Triangle, Rectangle, Square, Parallelogram, Trapezoid, Regular Polygon

**O(1) time/space for all**

```python
from math import sqrt, tan, pi

# Triangle (Heron's)
def triangle_area(a, b, c):
    s = (a + b + c) / 2
    return sqrt(s * (s - a) * (s - b) * (s - c))

def triangle_perimeter(a, b, c):
    return a + b + c

# Rectangle
def rectangle_area(length, width):
    return length * width

def rectangle_perimeter(length, width):
    return 2 * (length + width)

# Square
def square_area(side):
    return side ** 2

def square_perimeter(side):
    return 4 * side

# Parallelogram
def parallelogram_area(base, height):
    return base * height

def parallelogram_perimeter(base, side):
    return 2 * (base + side)

# Trapezoid
def trapezoid_area(a, b, height):
    return ((a + b) / 2) * height

def trapezoid_perimeter(a, b, c, d):
    return a + b + c + d

# Regular Polygon
def regular_polygon_area(n, side):
    return (n * side**2) / (4 * tan(pi / n))

def regular_polygon_perimeter(n, side):
    return n * side

print('Functions loaded')
```

## Example Calculations (matches C++)

```python
# Triangle 3-4-5
a, b, c = 3, 4, 5
print("Triangle (3,4,5):")
print("Area:", triangle_area(a,b,c))
print("Perimeter:", triangle_perimeter(a,b,c))

# Rectangle 5x10
print("\nRectangle (5,10):")
print("Area:", rectangle_area(5,10))
print("Perimeter:", rectangle_perimeter(5,10))

# Square side 4
print("\nSquare (4):")
print("Area:", square_area(4))
print("Perimeter:", square_perimeter(4))

# Parallelogram
print("\nParallelogram (base 6, height 3, side 5):")
print("Area:", parallelogram_area(6,3))
print("Perimeter:", parallelogram_perimeter(6,5))

# Trapezoid
print("\nTrapezoid (4,6 height 4):")
print("Area:", trapezoid_area(4,6,4))
print("Perimeter:", trapezoid_perimeter(4,6,3,5))

# Hexagon n=6 side 2
print("\nHexagon (n=6, side 2):")
print("Area:", regular_polygon_area(6,2))
print("Perimeter:", regular_polygon_perimeter(6,2))
```

## Interactive Calculator
Input dimensions for any shape.

```python
while True:
    print("\n1.Triangle 2.Rectangle 3.Square 4.Parallelogram 5.Trapezoid 6.Regular Polygon 0.Exit")
    choice = input("Shape: ").strip()
    
    if choice == '0': break
    
    if choice == '1':
        a,b,c = map(float, input("Sides a b c: ").split())
        print(f"Area: {triangle_area(a,b,c):.2f}, Perimeter: {triangle_perimeter(a,b,c):.2f}")
    elif choice == '2':
        l,w = map(float, input("Length width: ").split())
        print(f"Area: {rectangle_area(l,w):.2f}, Perimeter: {rectangle_perimeter(l,w):.2f}")
    elif choice == '3':
        s = float(input("Side: "))
        print(f"Area: {square_area(s):.2f}, Perimeter: {square_perimeter(s):.2f}")
    elif choice == '4':
        b,h = map(float, input("Base height: ").split())
        side = float(input("Side: "))
        print(f"Area: {parallelogram_area(b,h):.2f}, Perimeter: {parallelogram_perimeter(b,side):.2f}")
    elif choice == '5':
        a,b,h = map(float, input("Bases a b height: ").split())
        c,d = map(float, input("Legs c d: ").split())
        print(f"Area: {trapezoid_area(a,b,h):.2f}, Perimeter: {trapezoid_perimeter(a,b,c,d):.2f}")
    elif choice == '6':
        n = int(input("Sides n: "))
        s = float(input("Side length: "))
        print(f"Area: {regular_polygon_area(n,s):.2f}, Perimeter: {regular_polygon_perimeter(n,s):.2f}")
    else:
        print("Invalid")
```
