# Prime Number Checker - C++ to Python

**O(√n) time** - optimized check up to sqrt(n)

```python
from math import sqrt, isqrt

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, isqrt(num) + 1):
        if num % i == 0:
            return False
    return True

print('Prime checker loaded')
```

## Interactive Check

```python
while True:
    try:
        num = int(input("Enter number (0 to quit): "))
        if num == 0:
            break
        status = "prime" if is_prime(num) else "not prime"
        print(f"{num} is {status}")
    except:
        print("Enter valid integer")
```

## Test Cases

```python
test_nums = [1, 2, 17, 15, 97, 100]
for n in test_nums:
    print(f"{n}: {'prime' if is_prime(n) else 'not prime'}")

# Performance note: uses isqrt for O(√n)
```
