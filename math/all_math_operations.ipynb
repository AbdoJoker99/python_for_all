## Basic Arithmetic

```python
# + - * /
print(5 + 3)
print(10 - 2)
print(4 * 2)
print(16 / 2)  # float

# // % **
print(17 // 5)  # floor
print(17 % 5)   # remainder
print(3 ** 4)   # power
```

## Comparisons (from conditionals)

```python
age = 19
print(age == 18, age != 18)
print(age < 21, age > 21, age <= 21, age >= 21)
```

## Powers/Squares (lists, matplotlib)

```python
squares = [x**2 for x in range(1,11)]
print(squares)
```

```python
# Function math
def add_numbers(x, y):
    return x + y

print(add_numbers(5, 3))
```

## List Stats (min/max/sum)

```python
ages = [93, 99, 66, 17, 85, 1, 35, 82, 2, 77]
print(min(ages), max(ages), sum(ages), sum(ages)/len(ages))  # avg
```

```python
# Modulo for evens
numbers = list(range(1,11))
evens = [n for n in numbers if n % 2 == 0]
print(evens)
```
