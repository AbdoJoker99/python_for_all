# OOP Pillar 3: Polymorphism 🔄

**Definition:** Same interface, different implementations (override, duck typing, operator overload).

**Python:** Dynamic - no explicit interface, just implement methods.

**Roadmap:** Override → Duck typing → `@overload` → Operator → Examples.

## 1. Method Override (Runtime Polymorphism)

```python
class Bird:
    def sound(self):
        return 'Tweet'

class Rooster(Bird):
    def sound(self):  # override
        return 'Cock-a-doodle-doo!'

class Frog:
    def sound(self):  # unrelated class
        return 'Croak'

def make_sound(animal):  # polymorphic function
    print(animal.sound())

birds = [Bird(), Rooster()]
for b in birds:
    make_sound(b)
make_sound(Frog())
```

## 2. Duck Typing ('if it quacks like a duck')

```python
class Duck:
    def quack(self):
        return 'Quack!'

class Person:
    def quack(self):  # same method!
        return 'Quack quack! (imitation)'

def chorus(things):
    for t in things:
        print(t.quack())

chorus([Duck(), Person()])
```

## 3. Operator Overloading (__add__, __str__)

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f'Vector({self.x}, {self.y})'

v1 = Vector(2, 3)
v2 = Vector(1, 4)
print(v1 + v2)
```

## 4. Abstract Base Classes (Protocol)

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h

# s = Shape()  # TypeError
r = Rectangle(5, 3)
print(r.area())
```

## Examples Gallery

```python
# 5. MediaPlayer → Audio/Video
class MediaPlayer:
    def play(self, file):
        print(f'Playing {file}')

class AudioPlayer(MediaPlayer):
    def play(self, file):
        print(f'Audio: {file}')

class VideoPlayer(MediaPlayer):
    def play(self, file):
        print(f'Video: {file} (with visuals)')

players = [AudioPlayer(), VideoPlayer()]
for p in players:
    p.play('song.mp3')
```

## Function Overloading Simulation (default args)

```python
class Calculator:
    def add(self, a, b, c=None):
        if c is None:
            return a + b
        return a + b + c

calc = Calculator()
print(calc.add(1,2))      # 3
print(calc.add(1,2,3))    # 6
```
