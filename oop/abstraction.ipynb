# OOP Pillar 4: Abstraction 🎭

**Definition:** Hide complex implementation, show simple interface (abstract classes/interfaces).

**Python:** `ABC`, `@abstractmethod`, protocols.

**Roadmap:** ABC → Protocol → Examples → Factory pattern.

## 1. Abstract Base Class (ABC)

```python
from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardProcessor(PaymentProcessor):
    def pay(self, amount):
        print(f'Paid ${amount:.2f} with credit card')

class PayPalProcessor(PaymentProcessor):
    def pay(self, amount):
        print(f'Paid ${amount:.2f} via PayPal')

# processor = PaymentProcessor()  # TypeError!

processors = [CreditCardProcessor(), PayPalProcessor()]
for p in processors:
    p.pay(100)
```

## 2. Protocol (Structural Typing)

```python
from typing import Protocol

class Drawable(Protocol):
    def draw(self) -> None:
        ...

class Circle:
    def draw(self):
        print('Drawing circle')

class Square:
    def draw(self):
        print('Drawing square')

def render(shape: Drawable):
    shape.draw()

render(Circle())
render(Square())
```

## 3. Template Method Pattern

```python
class Game(ABC):
    def play(self):
        self.start()
        self.run_game()
        self.end()
    
    @abstractmethod
    def run_game(self):
        pass
    
    def start(self):
        print('Game started')
    
    def end(self):
        print('Game ended')

class Chess(Game):
    def run_game(self):
        print('Playing chess...')

class VideoGame(Game):
    def run_game(self):
        print('Playing video game...')

chess = Chess()
chess.play()
```

## Examples Gallery

```python
# 4. Database Connection
class Database(ABC):
    @abstractmethod
    def connect(self):
        pass

class MySQL(Database):
    def connect(self):
        print('Connected to MySQL')

class PostgreSQL(Database):
    def connect(self):
        print('Connected to PostgreSQL')

# 5. GUI Button (factory-like)
class Button(ABC):
    @abstractmethod
    def render(self):
        pass

class WindowsButton(Button):
    def render(self):
        print('Windows button')

class MacButton(Button):
    def render(self):
        print('Mac button')
```

## Factory Pattern (Abstraction + Creation)

```python
class ShapeFactory:
    @staticmethod
    def create_shape(type_):
        if type_ == 'circle':
            return Circle()
        elif type_ == 'square':
            return Square()
        raise ValueError('Unknown shape')

class Circle:
    def draw(self):
        print('Circle')

class Square:
    def draw(self):
        print('Square')

circle = ShapeFactory.create_shape('circle')
circle.draw()
```
