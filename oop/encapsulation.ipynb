# OOP Pillar 1: Encapsulation 🚀

**Definition:** Bundling data (attributes) and methods, restricting direct access to some components (data hiding).

**Python:** Convention `_private`, `__name_mangling`, `@property`.

**Learning Roadmap:** Basic → Access control → Properties → Validation → 10+ examples.

**Goal:** Protect object state, expose interfaces.

## 1. Basic Encapsulation (self)

```python
class Car:
    def __init__(self, make, model):
        self.make = make      # public
        self._model = model   # protected (convention)
    
    def get_model(self):      # getter
        return self._model
    
    def set_model(self, model):  # setter
        self._model = model

my_car = Car('Toyota', 'Camry')
print(my_car.make)
print(my_car.get_model())
my_car.set_model('Corolla')
print(my_car.get_model())
```

## 2. Private Attributes (`__` mangling)

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # private
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print('Invalid amount')
    
    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())
# print(account.__balance)  # AttributeError!
```

## 3. @property Decorator (Getter)

```python
class Person:
    def __init__(self, first, last):
        self.first = first
        self.last = last
    
    @property
    def full_name(self):
        return f'{self.first} {self.last}'
    
    @full_name.setter
    def full_name(self, name):
        self.first, self.last = name.split(' ', 1)

p = Person('John', 'Doe')
print(p.full_name)  # property access
p.full_name = 'Jane Smith'
print(p.first, p.last)
```

## 4. Validation in Setters

```python
class Temperature:
    def __init__(self, celsius=0):
        self._celsius = celsius
    
    @property
    def celsius(self):
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError('Too cold!')
        self._celsius = value
    
    @property
    def fahrenheit(self):
        return self._celsius * 9/5 + 32

temp = Temperature(25)
print(temp.fahrenheit)
# temp.celsius = -300  # ValueError
```

## Examples Gallery (10+)

```python
# 5. Rectangle area (protected)
class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height
    
    @property
    def area(self):
        return self._width * self._height

rect = Rectangle(5, 3)
print(rect.area)

# 6. Employee (private salary)
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary
    
    def give_raise(self, percent):
        self.__salary *= (1 + percent)
    
    def get_salary(self):
        return self.__salary

emp = Employee('Alice', 50000)
emp.give_raise(0.1)
print(emp.get_salary())
```
