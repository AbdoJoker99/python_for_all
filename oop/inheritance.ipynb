# OOP Pillar 2: Inheritance 👨‍👩‍👧

**Definition:** Child classes inherit attributes/methods from parent, extend/override.

**Python:** `class Child(Parent):`, `super()`.

**Roadmap:** Basic → super() → multilevel → override → 10+ examples.

## 1. Basic Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f'{self.name} makes a sound.')

class Dog(Animal):  # inherit
    pass

class Cat(Animal):
    pass

dog = Dog('Buddy')
cat = Cat('Whiskers')
dog.speak()
cat.speak()
```

## 2. Extend with super()

```python
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model
    
    def info(self):
        return f'{self.make} {self.model}'

class Car(Vehicle):
    def __init__(self, make, model, doors):
        super().__init__(make, model)  # parent constructor
        self.doors = doors
    
    def info(self):
        return f'{super().info()}, {self.doors} doors'

car = Car('Toyota', 'Camry', 4)
print(car.info())
```

## 3. Multilevel Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
class Mammal(Animal):
    def __init__(self, name, fur_color):
        super().__init__(name)
        self.fur_color = fur_color

class Dog(Mammal):
    def __init__(self, name, fur_color, breed):
        super().__init__(name, fur_color)
        self.breed = breed

my_dog = Dog('Buddy', 'brown', 'Labrador')
print(my_dog.name, my_dog.fur_color, my_dog.breed)
```

## 4. Method Override

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):  # override
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2

shapes = [Rectangle(5, 3), Circle(4)]
for shape in shapes:
    print(f'Area: {shape.area()}')
```

## Examples Gallery

```python
# 5. Employee → Manager
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, team_size):
        super().__init__(name, salary)
        self.team_size = team_size

mgr = Manager('Alice', 100000, 5)
print(mgr.name, mgr.team_size)

# 6. ElectricCar from original
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)
        self.battery_size = 70

ecar = ElectricCar('Tesla', 'Model S', 2023)
print(ecar.make, ecar.battery_size)
```

## Multiple Inheritance

```python
class Flyer:
    def fly(self):
        print('Flying!')

class Swimmer:
    def swim(self):
        print('Swimming!')

class Duck(Flyer, Swimmer):
    def quack(self):
        print('Quack!')

duck = Duck()
duck.fly()
duck.swim()
duck.quack()
```
