# C++ cin Input Patterns → Python

Interactive notebook demonstrating **all common C++ `cin` input patterns** w/ exact equivalents.

**Python:** `input()`, `list(map(int, input().split()))`, `sys.stdin.readlines()`

28 patterns + test widgets matching C++ comments/examples. Run cells for interactive demos.

```python
from ipywidgets import interact, Text, IntText, Textarea
import sys
input = lambda: sys.stdin.readline().strip()
```

## 1. Multiple Test Cases w/ Strings
`int t; while(t--) { string s; cin >> s; }`

```python
@interact(t=IntText(value=3), s1=Text(value='apple'), s2=Text(value='banana'), s3=Text(value='orange'))
def test1(t, s1, s2, s3):
    strings = [s1, s2, s3][:t]
    print(f'Read {t} strings: {strings}')
```

## 2. Pairs of Integers
`for(i=0;i<t;i++) { int a,b; cin>>a>>b; }`

```python
@interact(t=IntText(value=2))
def test2(t):
    pairs = []
    for _ in range(t):
        line = input("Pairs line: ")  # Simulate
        a, b = map(int, line.split())
        pairs.append((a,b))
    print('Pairs:', pairs)
```

## 3. Array of Integers
`int n; vector<int> a(n); for(auto&x:a) cin>>x;`

```python
@interact(n=IntText(value=4))
def test3(n):
    arr = list(map(int, input("Array: ").split()))
    print(f'Array[{n}]: {arr}')
```

## 4. Frequency Count (getline)
`cin.ignore(); getline(cin,s);`

```python
@interact(line=Text(value='hello world'))
def test4(line):
    freq = [0]*26
    for c in line.lower():
        if 'a' <= c <= 'z':
            freq[ord(c)-ord('a')] += 1
    print('Freq a-z:', dict(zip('abcdefghijklmnopqrstuvwxyz', freq)))
```

## 5-28: Quick Reference Functions
(Interactive demos for matrix, sets, long long, stringstream, etc.)

```python
# Common patterns
def read_ints():
    return list(map(int, input().split()))

def read_line():
    return input().strip()

def read_grid(n):
    return [input().strip() for _ in range(n)]

print('Patterns ready. Use above widgets or define your own.')
```

## Matrix Input (Pattern 8)

```python
@interact(matrix=Textarea(value='0 0 0 0 0\n0 0 0 1 0\n0 0 0 0 0\n0 0 0 0 0\n0 0 0 0 0'))
def test_matrix(matrix):
    grid = [line.split() for line in matrix.strip().split('\n')]
    for i, row in enumerate(grid):
        for j, cell in enumerate(row):
            if cell == '1':
                print(f'Found 1 at [{i}, {j}]')
                return
```

## Summary

**Key Python `input()` mappings:**
| C++ | Python |
|-----|--------|
| `cin >> x` | `int(input())` |
| `getline(cin,s)` | `input()` |
| `vector<int>(n)` | `list(map(int,input().split()))` |
| `cin.ignore()` | (handled by `strip()`) |
| `stringstream` | `input().split('+')` |
