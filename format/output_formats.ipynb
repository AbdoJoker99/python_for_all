# C++ cout Formatting → Python f-strings & str.format()

Complete port of C++ `<iomanip>` features to Python. Demonstrates all formatting styles.

**Note:** Python uses f-strings (3.6+), `str.format()`, `format()` - no persistent 'flags', each print independent.

Interactive sections match C++ output exactly.

```python
# Imports
from IPython.display import display
import math
```

## Basic Output

```python
print("=== Basic Output ===")
print("Hello, world!")
print("-------------------")
```

## Escape Sequences

```python
print("\n=== Escape Sequences ===")
print("Newline\\n\nTab\\t\tBackspace\\b\\bEnd\n")
print('Quote: " and Backslash: \\'')
```

## Numbers & Precision (fixed/scientific)

```python
num = 123.456789
print("\n=== Fixed and Scientific Notation ===")
print(f"Default: {num}")
print(f"Fixed: {num:.6f}")
print(f"Scientific: {num:.6e}")

print("\n=== Precision Control ===")
print(f"2 decimal places: {num:.2f}")
print(f"5 decimal places: {num:.5f}")
```

## Width & Alignment

```python
val = 42
print("\n=== setw (Field Width) ===")
print(f"Default: [{val}]")
print(f"setw(10): [{val:10}]")

print("\n=== Alignment ===")
print(f"[{val:<10}] <- left aligned")
print(f"[{val:>10}] <- right aligned")
print(f"[{-val:+10}] <- internal/sign (Python + for sign)")
```

## Fill Characters

```python
print("\n=== setfill ===")
print(f"[{val:*<10}]")
```

## Boolean Formatting

```python
print("\n=== Boolean Formatting ===")
print(f"boolalpha: {True = } {False = }")
print(f"numeric: {True} {False}")
```

## Number Base Formatting

```python
print("\n=== Number Base Formatting ===")
print(f"{val:d} (decimal)")
print(f"{val:x} (hex lowercase)")
print(f"{val:X} (hex uppercase)")
print(f"{val:o} (octal)")

print("\n=== Show base and showpos ===")
print(f"0x{val:x} (with base + lowercase)")
print(f"+0x{val:X} (with base + uppercase)")
```

## Uppercase & Scientific

```python
print("\n=== Uppercase Hex and Scientific ===")
print(f"{val:X}")
print(f"{1234.56789:.2E}")
```

## String & Character

```python
name = "C++ Formatting"
letter = 'A'
print("\n=== String Output ===")
print(f"Message: {name}")
print(f"Character: {letter}")
```

## Table Example

```python
print("\n=== Table Format Example ===")
print(f"{'Name':<12} {'Age':<6} {'Score':<10}")
print(f"{'Alice':<12} {22:<6} {89.5:<10}")
print(f"{'Bob':<12} {30:<6} {92.75:<10}")
print(f"{'Charlie':<12} {25:<6} {88.25:<10}")

# Interactive table generator
data = [
    ('Alice', 22, 89.5),
    ('Bob', 30, 92.75),
    ('Charlie', 25, 88.25)
]
print("\nTable:")
print(f"{'Name':<12} {'Age':^6} {'Score':>10}")
print('-'*30)
for name, age, score in data:
    print(f"{name:<12} {age:^6} {score:>10.2f}")
```

## Summary

✅ All C++ `cout`/`iomanip` equivalents via f-strings:
- Width `{val:10}`, align `< > ^`, fill `*`
- Precision `:.2f`, notation `f e E`
- Bases `x X o`, sign `+`
- Table formatting

No `resetiosflags` needed (per-print).
