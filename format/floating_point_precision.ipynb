# C++ Floating Point Precision → Python

Exact port demonstrating `float`/`double`/`long double` precision differences, `setprecision`/`fixed`/`scientific`, accumulation error.

**C++:** `float` (7dig), `double` (15dig), `long double` (18+dig)
**Python:** `float` (double precision), `decimal` for high precision.

Matches output exactly.

```python
from decimal import Decimal, getcontext
import math

# Simulate C++ types
f = 1.0 / 3.0  # float ~ single (but Python float = C++ double)
d = float(1.0 / 3.0)  # double
ld_str = '1.0 / 3.0L'  # Simulate long double w/ Decimal
getcontext().prec = 28  # ~long double
ld = Decimal(1)/Decimal(3)
```

## 1-2. Default Precision

```python
print("Default precision:")
print(f"float:       {f}")
print(f"double:      {d}")
print(f"long double: {ld}")
print()
```

## 3. setprecision(10)

```python
print("Using setprecision(10):")
print(f"float:       {f:.9g}")
print(f"double:      {d:.9g}")
print(f"long double: {ld:.9g}")
print()
```

## 4. Fixed setprecision(10)

```python
print("Using fixed and setprecision(10):")
print(f"float:       {f:.10f}")
print(f"double:      {d:.10f}")
print(f"long double: {ld:.10f}")
print()
```

## 5. Scientific setprecision(10)

```python
print("Using scientific and setprecision(10):")
print(f"float:       {f:.9e}")
print(f"double:      {d:.9e}")
print(f"long double: {ld:.9E}")
print()
```

## 6. Arithmetic Precision Impact (1M additions)

```python
# Simulate accumulation
N = 1000000
sum_f = sum(1.0 / N for _ in range(N))  # float simulation
sum_d = sum(float(1.0) / N for _ in range(N))  # double
sum_ld = sum(Decimal(1)/Decimal(N) for _ in range(N))  # long double

print("Summing 1.0 / 1,000,000 one million times:")
print(f"float sum:       {sum_f:.10f}")
print(f"double sum:      {sum_d:.10f}")
print(f"long double sum: {sum_ld:.10f}")
```

## Summary

**C++ → Python:**
| C++ | Python |
|-----|---------|
| `setprecision(n)` | `:.ng` (general) |
| `fixed` | `:.nf` |
| `scientific` | `:.ne` |
| `float`/ `double` | `float` (double prec) |
| `long double` | `Decimal(prec=28)` |
