# All Python Error Types

Syntax | Runtime | Logic - with handling.

## Syntax Errors

```python
# print('missing")  # Syntax
print('Syntax fixed')
```

## Runtime (Exceptions)

```python
try:
    5/0  # ZeroDivisionError
except ZeroDivisionError:
    print('caught')

try:
    [1][10]  # IndexError
except IndexError:
    print('caught')

try:
    int('abc')  # ValueError
except:
    print('caught')
```

## Logic Errors

```python
sum = 0
for i in range(5):
    sum -= i  # Wrong sign
print(sum)  # -10 not 10
```

## try/except/else/finally

```python
try:
    n = int('123')
except ValueError:
    print('bad')
else:
    print(n*n)
finally:
    print('end')
```
