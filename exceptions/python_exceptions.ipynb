# Handling Exceptions in Python


When you think your code might fail, you can use try-except blocks to handle errors (exceptions) gracefully instead of crashing the program.

## Basic try-exexcept

The `try` block runs code that might fail. The `except` block runs if an exception occurs.

```python
# This will cause an error
print(5/0)
```

```python
# Handling the error
try:
    print(5/0)
except:
    print("You can't divide by zero!")
```

## Specific Exceptions

Use specific exception types for better error handling.

```python
# Common exceptions
try:
    answer = 10 / int('abc')
except ZeroDivisionError:
    print("Can't divide by zero")
except ValueError:
    print("Invalid value")
else:
    print(f"Answer: {answer}")
```

## Multiple Exceptions in One except

Catch multiple exception types with tuples.

```python
def get_file_lines(filename):
    try:
        with open(filename, 'r') as f:
            return f.readlines()
    except (FileNotFoundError, PermissionError) as e:
        print(f"Error reading {filename}: {e}")
        return []
```

## else and finally Clauses

`else` runs if no exception. `finally` always runs (cleanup).

```python
try:
    number = int(input("Enter a number: "))
except ValueError:
    print("Not a number!")
else:
    print(f"Square is {number ** 2}")
finally:
    print("Execution complete.")
```

## Raising Exceptions

Use `raise` to create/throw your own exceptions.

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age can't be negative")
    if age < 18:
        raise PermissionError("Too young")
    print("Access granted")

# Test
try:
    check_age(-5)
except (ValueError, PermissionError) as e:
    print(f"Error: {e}")
```

## Summary

**Key concepts:**
- `try`/`except`: Handle errors gracefully
- Specific exceptions > generic `except:`
- `except (Type1, Type2)`: Multiple types
- `else`: No exception occurred
- `finally`: Always executes (cleanup)
- `raise`: Throw custom exceptions

**Practice:** Wrap risky code in try-except blocks!
