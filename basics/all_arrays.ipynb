## Array Basics (Access, Modify, Add/Remove)

```python
array = ['a', 'b', 'c']
print(array[0], array[-1])
array[0] = 'A'
array.append('d')
array.insert(1, 'X')
print(array)
del array[-1]
popped = array.pop(0)
print(popped, array)
```

## Sorting, Slicing, Copy

```python
numbers = [3,1,4,1,5]
numbers.sort()
print(numbers)
print(sorted(numbers, reverse=True))

slice_copy = numbers[:]
slice_copy.append(9)
print('Original:', numbers, 'Copy:', slice_copy)
```

## Array Comprehensions & Filtering

```python
squares = [x**2 for x in range(1,11)]
print(squares)
evens = [x for x in range(20) if x % 2 == 0]
print(evens)
```

```python
# Loop filter
arr = list(range(1,11))
evens_loop = []
for n in arr:
    if n % 2 == 0:
        evens_loop.append(n)
print(evens_loop)
```

## Array-like: range()

```python
arr_range = list(range(1,6))
print(arr_range)
print(list(range(0,10,2)))
```

## Nested Arrays (Matrices)

```python
matrix = [[1,2], [3,4]]
print(matrix[0][1])  # 2
```
