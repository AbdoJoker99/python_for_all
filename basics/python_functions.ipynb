# Functions in Python - Fixed Complete Edition

Defining, arguments, return values, lists, *args/**kwargs, modules.

## Defining a Function

```python
def greet_user():
    """Display a simple greeting."""
    print("Hello!")

greet_user()
```

## Arguments - Positional, Keyword, Default

```python
def describe_pet(animal, name='dog'):
    print(f"\nI have a {animal}.")
    print(f"My {animal}'s name is {name.title()}.")

describe_pet('hamster', 'harry')
describe_pet(name='willie')
```

## Return Values - Simple, Dict

```python
def build_person(first, last, age=None):
    person = {'first': first, 'last': last}
    if age:
        person['age'] = age
    return person

musician = build_person('jimi', 'hendrix', 27)
print(musician)
```

## Lists - Pass, Modify, Prevent Modify

```python
def print_models(unprinted, printed):
    while unprinted:
        current = unprinted.pop()
        print(f"Printing {current}")
        printed.append(current)

# Modify original
unprinted = ['phone case', 'pendant', 'ring']
printed = []
print_models(unprinted, printed)
print(f"Unprinted: {unprinted}")

# Prevent with copy [:]
original = ['phone case', 'pendant', 'ring']
printed = []
print_models(original[:], printed)
print(f"Original intact: {original}")
```

## Arbitrary Arguments *args, **kwargs

```python
def make_pizza(size, *toppings):
    print(f"Making {size} pizza with {len(toppings)} toppings:")
    [print(f" - {t}") for t in toppings]

make_pizza('small', 'pepperoni')
make_pizza('large', 'bacon', 'pineapple')

def build_profile(first, last, **info):
    profile = {'first': first, 'last': last}
    profile.update(info)
    return profile

print(build_profile('albert', 'einstein', location='princeton'))
```
