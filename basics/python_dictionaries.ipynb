# Python Dictionaries

Python's dictionaries allow you to connect pieces of related information. Each piece of information in a dictionary is stored as a **key-value pair**. When you provide a key, Python returns the value associated with that key.

---

## Defining a Dictionary

Use **curly braces** to define a dictionary. Use **colons** to connect keys and values, and use **commas** to separate individual key-value pairs.

```python
# Making a dictionary
alien_0 = {'color': 'green', 'points': 5}
print(alien_0)
```

---

## Accessing Values

To access the value associated with an individual key, give the name of the dictionary and then place the key in a set of square brackets.

**NOTE:** If the key you're asking for is not in the dictionary, an error will occur!

You can also use the `get()` method, which returns **None** instead of an error if the key doesn't exist.

```python
# Getting the value associated with a key
alien_0 = {'color': 'green', 'points': 5}
print(alien_0['color'])
print(alien_0['points'])
```

```python
# Getting the value with get() - safer approach!
alien_0 = {'color': 'green'}
alien_color = alien_0.get('color')
alien_points = alien_0.get('points', 0)  # Default value if key missing

print(alien_color)
print(alien_points)
```

**Tip:** Use `get(key, default_value)` to provide a fallback when keys don't exist!

---

## Adding New Key-Value Pairs

You can store as many key-value pairs as you want in a dictionary. To add a new key-value pair, give the name of the dictionary and the new key in square brackets, then set it equal to the new value.

```python
# Adding a key-value pair
alien_0 = {'color': 'green', 'points': 5}
alien_0['x'] = 0
alien_0['y'] = 25
alien_0['speed'] = 1.5
print(alien_0)
```

```python
# Starting with an empty dictionary
alien_0 = {}
alien_0['color'] = 'green'
alien_0['points'] = 5
print(alien_0)
```

---

## Modifying Values

You can modify the value associated with any key in a dictionary by giving the name of the dictionary and enclosing the key in square brackets, then providing the new value.

```python
# Modifying values in a dictionary
alien_0 = {'color': 'green', 'points': 5}
print(f"Before: {alien_0}")

# Change the alien's color and point value
alien_0['color'] = 'yellow'
alien_0['points'] = 10
print(f"After: {alien_0}")
```

---

## Removing Key-Value Pairs

You can remove any key-value pair you want from a dictionary using the `del` keyword.

```python
# Deleting a key-value pair
alien_0 = {'color': 'green', 'points': 5}
print(f"Before: {alien_0}")

del alien_0['points']
print(f"After: {alien_0}")
```

**Warning:** Once you delete a key-value pair, it's gone forever!

---

## Looping Through a Dictionary

You can loop through a dictionary in three ways:
1. All the **key-value pairs**
2. All the **keys**
3. All the **values**

**Note:** A dictionary only tracks the connections between keys and values; it doesn't track the order of items. Use `sorted()` if you need ordered results!

```python
# Looping through all key-value pairs
fav_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

for name, language in fav_languages.items():
    print(f"{name}: {language}")
```

```python
# Looping through all the keys
fav_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

for name in fav_languages.keys():
    print(name)
```

```python
# Looping through all the values
fav_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

for language in fav_languages.values():
    print(language)
```

```python
# Looping through all the keys in order (sorted)
fav_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

for name in sorted(fav_languages.keys()):
    print(f"{name}: {fav_languages[name]}")
```

---

## Dictionary Length

You can find the number of key-value pairs in a dictionary using `len()`.

```python
# Finding a dictionary's length
fav_languages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
}

num_responses = len(fav_languages)
print(f"Number of responses: {num_responses}")
```
