# All Comment Types Filtered from Project Files

Python comments: # single-line, """triple-double quotes""" docstrings, '''triple-single quotes'''.

## # Single Line Comments

```python
# This is a single line comment
# Comments ignored by interpreter
x = 5  # Inline comment
print(x)  # Print value
```

## """ Triple Double Quotes - Docstrings (functions/classes)

```python
def greet_user():
    """Display a simple greeting."""
    print("Hello!")

greet_user()
print(greet_user.__doc__)  # Access docstring
```

```python
class Car:
    """A simple attempt to model a car."""
    def __init__(self, make):
        """Initialize car attributes."""
        self.make = make

print(Car.__doc__)
```

## ''' Triple Single Quotes - Multi-line Docstrings

```python
def make_pizza():
    '''Make a pizza.
    Multi-line description.
    '''
    print("Pizza ready")

print(make_pizza.__doc__)
```

## Multi-line Block Comments (convention)

```python
'''
Multi-line comment block
Used when triple quotes not for docstrings
'''
print("Code after block comment")
```

## Examples from Project Files

(Filtered docstrings & # comments)

## Best Practices

- # for single line/explanation
- """ for function/class/module docstrings (PEP 257)
- help(func) or func.__doc__ to view

```python
help(describe_pet)
```

## Summary
3 comment types: #, """docstring""", '''multi'''. Use docstrings for self-documenting code.
