# If Statements and While Loops

If statements allow you to examine the current state of a program and respond appropriately. While loops run as long as certain conditions remain true.

---

## Conditional Tests

A **conditional test** is an expression that can be evaluated as True or False. Python uses these values to decide whether the code in an if statement should be executed.

```python
# Checking for equality
# A single equal sign (=) assigns a value to a variable.
# A double equal sign (==) checks whether two values are equal.

car = 'bmw'
print(car == 'bmw')   # True
print(car == 'audi')  # False
```

```python
# Ignoring case when making a comparison

car = 'Audi'
print(car.lower() == 'audi')  # True
```

```python
# Checking for inequality

topping = 'mushrooms'
print(topping != 'anchovies')  # True
```

---

## Numerical Comparisons

Testing numerical values is similar to testing string values.

```python
# Testing equality and inequality

age = 18
print(f"age = {age}")
print(f"age == 18: {age == 18}")   # True
print(f"age != 18: {age != 18}")   # False
```

```python
# Comparison operators

age = 19
print(f"age = {age}")
print(f"age < 21: {age < 21}")      # True
print(f"age <= 21: {age <= 21}")    # True
print(f"age > 21: {age > 21}")      # False
print(f"age >= 21: {age >= 21}")    # False
```

---

## Checking Multiple Conditions

You can check multiple conditions at the same time. The **and** operator returns True if all the conditions listed are True. The **or** operator returns True if any condition is True.

```python
# Using and to check multiple conditions

age_0 = 22
age_1 = 18
print(f"age_0 = {age_0}, age_1 = {age_1}")
print(f"age_0 >= 21 and age_1 >= 21: {age_0 >= 21 and age_1 >= 21}")  # False
```

```python
# Using or to check multiple conditions

age_0 = 22
age_1 = 18
print(f"age_0 = {age_0}, age_1 = {age_1}")
print(f"age_0 >= 21 or age_1 >= 21: {age_0 >= 21 or age_1 >= 21}")  # True
```

---

## Boolean Values

A **boolean value** is either True or False. Variables with boolean values are often used to keep track of certain conditions within a program.

```python
# Boolean values

game_active = True
can_edit = False

print(f"game_active: {game_active}")
print(f"can_edit: {can_edit}")
```

---

## If Statements

Several kinds of if statements exist. Your choice depends on the number of conditions you need to test.

```python
# Simple if statement

age = 19
if age >= 18:
    print("You're old enough to vote!")
```

```python
# If-else statements

age = 17
if age >= 18:
    print("You're old enough to vote!")
else:
    print("You can't vote yet.")
```

```python
# The if-elif-else chain

age = 12
if age < 4:
    price = 0
elif age < 18:
    price = 5
else:
    price = 10

print(f"Your cost is ${price}.")
```

---

## Conditional Tests with Lists

You can easily test whether a certain value is in a list or whether a list is empty.

```python
# Testing if a value is in a list

players = ['al', 'bea', 'cyn', 'dale']

print(f"'al' in players: {'al' in players}")      # True
print(f"'eric' in players: {'eric' in players}")  # False
```
