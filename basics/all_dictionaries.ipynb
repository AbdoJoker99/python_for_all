## Basic Dict (python_dictionaries.ipynb, beginners_python.ipynb)

```python
alien = {'color': 'green', 'points': 5}
print(alien)
print(alien['color'])
```

```python
# Access with get()
alien_color = alien.get('color')
alien_points = alien.get('points', 0)  # Default
print(alien_color, alien_points)
```

```python
# Add/Modify
alien['x'] = 0
alien['color'] = 'yellow'
print(alien)

# Del
del alien['points']
print(alien)
```

## Looping Dicts

```python
fav_languages = {'jen': 'python', 'sarah': 'c', 'phil': 'python'}

# items()
for name, lang in fav_languages.items():
    print(f"{name}: {lang}")

# keys()
for name in fav_languages.keys():
    print(name)

# values()
for lang in fav_languages.values():
    print(lang)

# sorted keys
for name in sorted(fav_languages.keys()):
    print(name)
```

```python
print(len(fav_languages))
```

## From Functions (build_person)

```python
def build_person(first, last, age=None):
    person = {'first': first, 'last': last}
    if age:
        person['age'] = age
    return person

p = build_person('albert', 'einstein', 42)
print(p)
```

## Nesting (lists in dicts, dicts in lists)

```python
# Lists in dict
langs = {'jen': ['python', 'ruby'], 'sarah': ['c']}
for name, lang_list in langs.items():
    print(name)
    for lang in lang_list:
        print('  ', lang)

# List of dicts
users = [
    {'first': 'albert', 'last': 'einstein'},
    {'first': 'marie', 'last': 'curie'}
]
for u in users:
    print(u)
```
