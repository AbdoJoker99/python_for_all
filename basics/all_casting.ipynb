## From exceptions demos

```python
try:
    number = int("abc")  # ValueError
except ValueError:
    print("Invalid int conversion")

try:
    num_tickets = int("3.5")  # ValueError
except ValueError:
    print("Please enter integer")
```

```python
# Successful input casting
prompt = "Enter number of tickets: "
# num_tickets = input(prompt)  # Simulate
num_tickets = "5"
try:
    num_tickets = int(num_tickets)
    print(f"Tickets: {num_tickets}")
except ValueError:
    print("Invalid!")
else:
    print("Printing tickets...")
```
