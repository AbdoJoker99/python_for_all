## 🎯 The range() Function

You can use the `range()` function to work with a set of numbers efficiently. It starts at 0 by default, and stops one number below the number passed to it.

````python
# Printing the numbers 0 to 1000
# Note: range stops at 1000 (exclusive), so we use 1001
for number in range(6):  # Just showing first 6 for brevity
    print(number)
````

````python
# Printing the numbers 1 to 1000 (start, stop)
for number in range(1, 6):  # Showing first 6 for brevity
    print(number)
````

````python
# Making a list of numbers using list()
numbers = list(range(1, 11))  # 1 to 10
print(numbers)
````

⚡ **range() with step:** You can add a third argument for step!
```python
list(range(0, 10, 2))  # [0, 2, 4, 6, 8]
list(range(5, 0, -1))  # [5, 4, 3, 2, 1]
```

## 📊 Simple Statistics

Python provides built-in functions for basic statistical analysis on numerical lists.

````python
# Sample data
ages = [93, 99, 66, 17, 85, 1, 35, 82, 2, 77]
print(f"Ages: {ages}")
````

````python
# Finding the minimum value
youngest = min(ages)
print(f"Youngest: {youngest}")
````

````python
# Finding the maximum value
oldest = max(ages)
print(f"Oldest: {oldest}")
````

````python
# Finding the sum of all values
total_years = sum(ages)
print(f"Total years: {total_years}")
````

💡 **Quick stats:** Combine all three for a quick summary!
```python
print(min(ages), max(ages), sum(ages))
```

## ✂️ Slicing a List

You can work with any set of elements from a list. A portion of a list is called a **slice**.

````python
# Sample list for slicing
finishers = ['kai', 'abe', 'ada', 'gus', 'zoe']
print(f"All finishers: {finishers}")
````

````python
# Getting the first three items (start:stop, stop is exclusive)
first_three = finishers[:3]
print(f"First three: {first_three}")
````

````python
# Getting the middle three items (index 1 to 4)
middle_three = finishers[1:4]
print(f"Middle three: {middle_three}")
````

````python
# Getting the last three items using negative index
last_three = finishers[-3:]
print(f"Last three: {last_three}")
````

💡 **Slice shorthand:** Leave first index empty = start from beginning. Leave last index empty = go to end.

## 📋 Copying a List

To copy a list, make a slice that starts at the first item and ends at the last item (`[:]`). If you try to copy a list without using this approach, whatever you do to the copied list will affect the original list as well!

````python
# Making a copy of a list using slicing
finishers = ['kai', 'abe', 'ada', 'gus', 'zoe']
copy_of_finishers = finishers[:]

print(f"Original: {finishers}")
print(f"Copy: {copy_of_finishers}")

# Modify the copy
copy_of_finishers.append('alex')
print(f"After modifying copy:")
print(f"Original: {finishers}")  # Unchanged!
print(f"Copy: {copy_of_finishers}")
````

⚠️ **Important:** Don't use `copy = original` - this creates a reference, not a copy!
```python
# WRONG - both point to same list!
copy = original

# CORRECT - creates a new copy
copy = original[:]
```

## ⚡ List Comprehensions

A **list comprehension** allows you to generate a list in one line of code. It combines a for loop and the creation of new elements into one compact syntax.

````python
# Using a loop to generate a list of square numbers
squares = []
for x in range(1, 11):
    square = x**2
    squares.append(square)
print(squares)
````

````python
# Using a comprehension to generate a list of square numbers
squares = [x**2 for x in range(1, 11)]
print(squares)
````

````python
# Using a comprehension to convert a list of names to upper case
names = ['kai', 'abe', 'ada', 'gus', 'zoe']
upper_names = [name.upper() for name in names]
print(upper_names)
````

📝 **Comprehension syntax:**
```python
[expression for item in iterable if condition]
```

## 🔒 Tuples

A **tuple** is like a list, except you **can't change** the values in a tuple once it's defined. Tuples are designated by **parentheses** instead of square brackets. Use tuples for data that shouldn't change!

````python
# Defining a tuple
dimensions = (800, 600)
print(f"Screen resolution: {dimensions[0]} x {dimensions[1]}")
````

````python
# Looping through a tuple
dimensions = (800, 600)
for dimension in dimensions:
    print(dimension)
````

````python
# Overwriting a tuple (You can't change elements, but you can assign a new value)
dimensions = (800, 600)
print(f"Original: {dimensions}")

dimensions = (1200, 900)
print(f"Overwritten: {dimensions}")
````

💡 **When to use tuples:**
- Coordinates (x, y)
- RGB colors (255, 0, 0)
- Configuration values that shouldn't change
- Dictionary keys (tuples are hashable)
