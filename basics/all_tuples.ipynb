## Defining & Accessing Tuples (python_advanced_lists.ipynb)

```python
dimensions = (800, 600)
print(f"Original: {dimensions}")
print(dimensions[0], dimensions[1])
```

```python
# Loop through tuple
for dimension in dimensions:
    print(dimension)
```

```python
# Immutability - Can't modify elements
# dimensions[0] = 1200  # ERROR!

# But can reassign whole tuple
dimensions = (1200, 900)
print(f"Reassigned: {dimensions}")
```
