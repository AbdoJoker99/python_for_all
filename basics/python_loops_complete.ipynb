## 🔄 For Loops

Iterate over sequences (lists, range, dicts).

```python
# Basic for loop over list (from python_lists)
users = ['val', 'bob', 'mia']
for user in users:
    print(f"Hello, {user.title()}")
print("Done!")

# With range (from advanced_lists)
for number in range(1, 6):
    print(number)
```

## ♾️ While Loops

Repeat while condition true.

```python
# Simple while (from intermediate_python)
current = 1
while current <= 5:
    print(current)
    current += 1

# User input while (conditionals_loops inspired)
# msg = ''
# while msg != 'quit':
#     msg = input('Message: ')
#     if msg:
#         print(msg)
```

## 🚦 Loop Controls: break, continue, pass

```python
# break
for num in range(10):
    if num == 5:
        break
    print(num)

# continue
for num in range(5):
    if num % 2 == 0:
        continue
    print(f"Odd: {num}")

# pass (placeholder)
while True:
    pass  # TODO
```

## 🔍 Filtering with Loops

Manual filtering (compare to list comps).

```python
# Filter evens with for
numbers = list(range(1, 11))
evens = []
for n in numbers:
    if n % 2 == 0:
        evens.append(n)
print("For evens:", evens)

# While filter
i = 0
positives = []
while i < len(numbers):
    if numbers[i] > 0:
        positives.append(numbers[i])
    i += 1
print("While positives:", positives)
```

## 🥚 Nested Loops (from nesting)

Loop in loop.

```python
# List of dicts
users = [
    {'name': 'albert', 'lang': 'python'},
    {'name': 'marie', 'lang': 'c'}
]
for user in users:
    for key, val in user.items():
        print(f"{key}: {val}")

# Dicts with lists
langs = {'jen': ['python', 'ruby']}
for name, lang_list in langs.items():
    print(name)
    for lang in lang_list:
        print(f"  - {lang}")
```

## 💡 Loops with If (from conditionals_loops)

Common pattern.

```python
votes = [18, 25, 16, 30]
can_vote = []
for age in votes:
    if age >= 18:
        can_vote.append(age)
print(can_vote)
```
