## 🔄 While Loops

A **while loop** repeats a block of code as long as a certain condition is true.

```python
# A simple while loop
current_value = 1
while current_value <= 5:
    print(current_value)
    current_value += 1
```

```python
# Letting the user choose when to quit
msg = ''
while msg != 'quit':
    msg = input("What's your message? ")
    print(msg)
```

💡 **Tip:** Use `break` to exit a loop immediately or `continue` to skip to the next iteration.

## 🛠️ Functions

**Functions** are named blocks of code, designed to do one specific job. Information passed to a function is called an **argument**, and information received by a function is called a **parameter**.

```python
# A simple function
def greet_user():
    """Display a simple greeting."""
    print("Hello!")

greet_user()
```

```python
# Passing an argument
def greet_user(username):
    """Display a personalized greeting."""
    print(f"Hello, {username}!")

greet_user('jesse')
```

```python
# Default values for parameters
def make_pizza(topping='bacon'):
    """Make a single-topping pizza."""
    print(f"Have a {topping} pizza!")

make_pizza()
make_pizza('pepperoni')
```

```python
# Returning a value
def add_numbers(x, y):
    """Add two numbers and return the sum."""
    return x + y

sum_result = add_numbers(3, 5)
print(sum_result)
```

⚡ **Pro Tip:** Use *args for multiple arguments and **kwargs for keyword arguments!

## 🐕 Classes

A **class** defines the behavior of an object and the kind of information an object can store. Information in a class is stored in **attributes**, and functions that belong to a class are called **methods**. A **child class** inherits the attributes and methods from its **parent class**.

```python
# Creating a dog class
class Dog():
    """Represent a dog."""
    
    def __init__(self, name):
        """Initialize dog object."""
        self.name = name
    
    def sit(self):
        """Simulate sitting."""
        print(f"{self.name} is sitting.")

my_dog = Dog('Peso')
print(f"{my_dog.name} is a great dog!")
my_dog.sit()
```

```python
# Inheritance - Creating a child class
class SARDog(Dog):
    """Represent a search dog."""
    
    def __init__(self, name):
        """Initialize the sardog."""
        super().__init__(name)
    
    def search(self):
        """Simulate searching."""
        print(f"{self.name} is searching.")

my_dog = SARDog('Willie')
print(f"{my_dog.name} is a search dog.")
my_dog.sit()
my_dog.search()
```

💡 **Tip:** The `super()` function lets the child class inherit from its parent class!

## 📁 Working with Files

Your programs can **read from** and **write to** files. Files are opened in **read mode** ('r') by default, but can also be opened in **write mode** ('w') and **append mode** ('a').

```python
# Reading a file and storing its lines
# Note: This will work if 'siddhartha.txt' exists
filename = 'siddhartha.txt'
try:
    with open(filename) as file_object:
        lines = file_object.readlines()
    for line in lines:
        print(line)
except FileNotFoundError:
    print(f"File '{filename}' not found. Creating a demo file...")
    # Create a demo file for testing
    with open(filename, 'w') as file_object:
        file_object.write("The Buddha, or Siddhartha Gautama, achieved enlightenment around 528 BC.")
    print("Demo file created!")
```

```python
# Writing to a file
filename = 'journal.txt'
with open(filename, 'w') as file_object:
    file_object.write("I love programming.")
print(f"Written to {filename}")
```

```python
# Appending to a file
filename = 'journal.txt'
with open(filename, 'a') as file_object:
    file_object.write("\nI love making games.")
print(f"Appended to {filename}")
```

⚠️ **Important:** Use 'w' to overwrite and 'a' to append!

## ⚠️ Exceptions

**Exceptions** help you respond appropriately to errors. Place code that might cause an error in the **try** block. Code that should run in response to an error goes in the **except** block. Code that should run only if the try block was successful goes in the **else** block.

```python
# Catching an exception
prompt = "How many tickets do you need? "
num_tickets = input(prompt)

try:
    num_tickets = int(num_tickets)
except ValueError:
    print("Please try again.")
else:
    print("Your tickets are printing.")
```

💡 **Common Exceptions:**
- `ValueError` - Wrong type conversion
- `FileNotFoundError` - File doesn't exist
- `ZeroDivisionError` - Division by zero
- `IndexError` - List index out of range
