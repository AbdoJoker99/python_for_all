# 📋 Python Lists - Complete Guide

A list stores a series of items in a particular order. Lists allow you to store sets of information in one place, whether you have just a few items or millions of items. Lists are one of Python's most powerful features readily accessible to new programmers!

---

## 📝 Defining a List

Use **square brackets** to define a list, and use **commas** to separate individual items in the list. Use **plural names** for lists, to make your code easier to read.

````python
# Making a list
users = ['val', 'bob', 'mia', 'ron', 'ned']
print(users)
````

---

## 🔍 Accessing Elements

Individual elements in a list are accessed according to their position, called the **index**. The index of the first element is **0**, the index of the second element is **1**, and so forth. Negative indices refer to items at the end of the list.

````python
# Getting the first element (index 0)
users = ['val', 'bob', 'mia', 'ron', 'ned']
first_user = users[0]
print(f"First user: {first_user}")
````

````python
# Getting the second element (index 1)
second_user = users[1]
print(f"Second user: {second_user}")
````

````python
# Getting the last element using -1
newest_user = users[-1]
print(f"Last user: {newest_user}")
````

💡 **Tip:** Use `[-2]` for second-to-last, `[-3]` for third-to-last, etc.

---

## ✏️ Modifying Individual Items

Once you've defined a list, you can change individual elements in the list by referring to their index.

````python
# Changing an element
users = ['val', 'bob', 'mia', 'ron', 'ned']
users[0] = 'valerie'
users[-2] = 'ronald'
print(users)
````

---

## ➕ Adding Elements

You can add elements to the end of a list, or you can insert them wherever you like in a list.

````python
# Adding an element to the end of the list
users = ['val', 'bob', 'mia', 'ron', 'ned']
users.append('amy')
print(users)
````

````python
# Starting with an empty list
users = []
users.append('val')
users.append('bob')
users.append('mia')
print(users)
````

````python
# Inserting elements at a particular position
users = ['val', 'bob', 'mia', 'ron', 'ned']
users.insert(0, 'joe')    # Insert at the beginning
users.insert(3, 'bea')    # Insert at index 3
print(users)
````

⚡ **insert() vs append():** Use `insert()` to add at a specific position, `append()` to add at the end.

---

## ➖ Removing Elements

You can remove elements by their **position** in a list, or by the **value** of the item.

````python
# Deleting an element by its position
users = ['val', 'bob', 'mia', 'ron', 'ned']
del users[-1]  # Delete the last element
print(users)
````

````python
# Removing an item by its value
users = ['val', 'bob', 'mia', 'ron', 'ned']
users.remove('mia')  # Removes only the first occurrence
print(users)
````

⚠️ **Note:** If the value appears multiple times, `remove()` only removes the first one!

---

## 🎯 Popping Elements

If you want to work with an element that you're removing from the list, you can **"pop"** the element. Think of the list as a stack of items - `pop()` takes an item off the top. By default it returns the last element, but you can pop from any position.

````python
# Pop the last item from a list
users = ['val', 'bob', 'mia', 'ron', 'ned']
most_recent_user = users.pop()
print(f"Popped: {most_recent_user}")
print(f"Remaining: {users}")
````

````python
# Pop the first item in a list
users = ['val', 'bob', 'mia', 'ron', 'ned']
first_user = users.pop(0)
print(f"Popped: {first_user}")
print(f"Remaining: {users}")
````

💡 **When to use pop() vs del:** Use `pop()` when you need the value, use `del` when you don't.

---

## 📏 List Length

The `len()` function returns the number of items in a list.

````python
# Find the length of a list
users = ['val', 'bob', 'mia', 'ron', 'ned']
num_users = len(users)
print(f"We have {num_users} users.")
````

---

## 🔃 Sorting a List

The `sort()` method changes the order of a list **permanently**. The `sorted()` function returns a copy of the list, leaving the original list **unchanged**.

````python
# Sorting a list permanently
users = ['val', 'bob', 'mia', 'ron', 'ned']
users.sort()
print(users)
````

````python
# Sorting a list permanently in reverse alphabetical order
users = ['val', 'bob', 'mia', 'ron', 'ned']
users.sort(reverse=True)
print(users)
````

````python
# Sorting a list temporarily (original unchanged)
users = ['val', 'bob', 'mia', 'ron', 'ned']
print("Original:", users)
print("Sorted:", sorted(users))
print("Sorted reverse:", sorted(users, reverse=True))
print("After sorted():", users)  # Original unchanged!
````

````python
# Reversing the order of a list
users = ['val', 'bob', 'mia', 'ron', 'ned']
users.reverse()
print(users)
````

⚠️ **Case sensitivity:** `sort()` treats uppercase and lowercase differently. Consider using `key=str.lower` for case-insensitive sorting:
```python
users.sort(key=str.lower)
```

---

## 🔁 Looping Through a List

Lists can contain millions of items, so Python provides an efficient way to loop through all the items. When you set up a loop, Python pulls each item from the list one at a time and stores it in a temporary variable.

````python
# Printing all items in a list
users = ['val', 'bob', 'mia', 'ron', 'ned']
for user in users:
    print(user)
````

````python
# Printing a message for each item, and a separate message afterwards
users = ['val', 'bob', 'mia', 'ron', 'ned']
for user in users:
    print(f"Welcome, {user}!")
print("Welcome, we're glad to see you all!")
````

💡 **Tip:** Use singular names (user) for the loop variable when iterating over plural names (users)!
