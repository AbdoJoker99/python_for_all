## 📦 Variables and Strings

**Variables** are used to store values. Think of them as containers with labels.

A **string** is a series of characters, surrounded by single or double quotes.

````python
# Hello World - Your first program!
print("Hello world!")
````

````python
# Hello World with a variable
msg = "Hello world!"
print(msg)
````

````python
# Concatenation - Combining strings
first_name = 'albert'
last_name = 'einstein'
full_name = first_name + ' ' + last_name
print(full_name)
````

💡 **Tip:** You can use `f-strings` for cleaner string formatting:
```python
print(f"{first_name} {last_name}")
```

## 📋 Lists

A **list** stores a series of items in a particular order. You access items using an index (starting from 0), or within a loop.

````python
# Make a list
bikes = ['trek', 'redline', 'giant']
print(bikes)
````

````python
# Get the first item in a list (index 0)
first_bike = bikes[0]
print(f"First bike: {first_bike}")
````

````python
# Get the last item in a list using -1
last_bike = bikes[-1]
print(f"Last bike: {last_bike}")
````

````python
# Looping through a list
print("All bikes:")
for bike in bikes:
    print(f"  - {bike}")
````

````python
# Adding items to a list
bikes = []  # Empty list
bikes.append('trek')
bikes.append('redline')
bikes.append('giant')
print(bikes)
````

````python
# Making numerical lists
squares = []
for x in range(1, 11):
    squares.append(x**2)
print(squares)
````

⚡ **List Comprehensions** - A more Pythonic way to create lists!

````python
# List comprehensions - One-liner way!
squares = [x**2 for x in range(1, 11)]
print(squares)
````

````python
# Slicing a list - Get portions of a list
finishers = ['sam', 'bob', 'ada', 'bea']
first_two = finishers[:2]
print(f"First two finishers: {first_two}")
````

````python
# Copying a list
copy_of_bikes = bikes[:]
print(f"Copy of bikes: {copy_of_bikes}")
````

## 🔒 Tuples

**Tuples** are similar to lists, but the items in a tuple **can't be modified** (they're immutable). Use them for fixed data.

````python
# Making a tuple - Note the parentheses instead of brackets
dimensions = (1920, 1080)
print(f"Screen resolution: {dimensions[0]} x {dimensions[1]}")
````

## 🔀 If Statements

**If statements** are used to test for particular conditions and respond appropriately.

### Conditional Tests:
- `x == y` - equals
- `x != y` - not equal
- `x > y` - greater than
- `x >= y` - greater than or equal to
- `x < y` - less than
- `x <= y` - less than or equal to

````python
# Conditional test with lists
bikes = ['trek', 'redline', 'giant']

print("'trek' in bikes:", 'trek' in bikes)
print("'surly' not in bikes:", 'surly' not in bikes)
````

````python
# Assigning boolean values
game_active = True
can_edit = False

print(f"Game is active: {game_active}")
print(f"Can edit: {can_edit}")
````

````python
# A simple if test
age = 19
if age >= 18:
    print("You can vote!")
````

````python
# If-elif-else statements
age = 15

if age < 4:
    ticket_price = 0
elif age < 18:
    ticket_price = 10
else:
    ticket_price = 15

print(f"Ticket price: ${ticket_price}")
````

## 📖 Dictionaries

**Dictionaries** store connections between pieces of information. Each item is a **key-value pair**.

````python
# A simple dictionary
alien = {'color': 'green', 'points': 5}
print(alien)
````

````python
# Accessing a value
print("The alien's color is " + alien['color'])
````

````python
# Adding a new key-value pair
alien['x_position'] = 0
print(alien)
````

````python
# Looping through all key-value pairs
fav_numbers = {'eric': 17, 'ever': 4}
for name, number in fav_numbers.items():
    print(f"{name} loves {number}")
````

````python
# Looping through all keys
for name in fav_numbers.keys():
    print(f"{name} loves a number")
````

````python
# Looping through all the values
for number in fav_numbers.values():
    print(f"{number} is a favorite")
````

## ⌨️ User Input

Your programs can **prompt the user for input**. All input is stored as a string!

````python
# Prompting for a value
name = input("What's your name? ")
print(f"Hello, {name}!")
````

````python
# Prompting for numerical input - Need to convert!
age = input("How old are you? ")
age = int(age)  # Convert string to integer
print(f"You are {age} years old.")
````

````python
# Working with floats
pi = input("What's the value of pi? ")
pi = float(pi)  # Convert string to float
print(f"Pi is approximately {pi}")
````
