## Basic Lists (python_lists.ipynb)

```python
# Define & Access
users = ['val', 'bob', 'mia', 'ron', 'ned']
print(users[0], users[-1])

# Modify
users[0] = 'valerie'
print(users)

# Add
users.append('amy')
users.insert(0, 'joe')
print(users)

# Remove
del users[-1]
users.remove('bob')
popped = users.pop(0)
print(popped, users, len(users))
```

```python
# Sort
cars = ['bmw', 'audi', 'subaru']
cars.sort()
print('Sorted:', cars)
print('Temp:', sorted(cars, reverse=True))
```

## Advanced Lists (python_advanced_lists.ipynb)

```python
# range
print(list(range(1,6)))
print(list(range(0,10,2)))

# Stats
ages = [93,99,66,17,85,1,35,82,2,77]
print(min(ages), max(ages), sum(ages))
```

```python
# Slicing & Copy
finishers = ['kai', 'abe', 'ada', 'gus', 'zoe']
print(finishers[:3])
print(finishers[1:4])
print(finishers[-3:])

copy = finishers[:]
copy.append('alex')
print('Original:', finishers)
print('Copy:', copy)
```

## List Comprehensions & Filtering

```python
# Comprehensions
squares = [x**2 for x in range(1,11)]
print(squares)

names = ['kai','abe','ada']
upper = [n.upper() for n in names]
print(upper)

# Filter
numbers = list(range(1,21))
evens = [x for x in numbers if x % 2 == 0]
print(evens)
```

## Filtering in Loops (python_loops_complete.ipynb)

```python
numbers = list(range(1,11))
evens = []
for n in numbers:
    if n % 2 == 0:
        evens.append(n)
print(evens)
```

## Nesting (python_nesting.ipynb, python_lists_complete.ipynb)

```python
# List of dicts
users = [
    {'first': 'albert', 'last': 'einstein'},
    {'first': 'marie', 'last': 'curie'}
]
for u in users:
    print(u['first'], u['last'])

# Lists in dict
langs = {'jen': ['python', 'ruby']}
for name, lang_list in langs.items():
    print(name)
    for lang in lang_list:
        print('  -', lang)
```
